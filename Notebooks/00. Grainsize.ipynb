# Theoretical transport reduction

```python
def get_transport_reduction(u, f,
                            A=100., rho_p=2650., rho_a=1.25, g=9.81,
                            d_n=np.arange(200e-6, 455e-6, 10e-6)):
    
    f = np.sqrt(f)
    d_n, u = np.meshgrid(d_n, u)

    uth_1 = A * np.sqrt((rho_p-rho_a)/rho_p*g*d_n)
    uth_2 = A * np.sqrt((rho_p-rho_a)/rho_p*g*f*d_n)

    q_1 = np.maximum(0., u - uth_1)**3
    q_2 = np.maximum(0., u - uth_2)**3

    r = (1 - q_2 / q_1) * 100.
    r[np.isnan(r)] = 100.
    
    return r


def lines_to_polygon(u, r):

    x = np.concatenate((u[:], u[-1:0:-1]))
    y = np.concatenate((np.min(r, axis=1),
                        np.max(r, axis=1)[-1:0:-1]))
    xy = np.concatenate((x[:,np.newaxis], y[:,np.newaxis]), axis=1)
    
    return xy


u = np.arange(0., 41., .1)

fig, axs = plt.subplots()

for i, f in enumerate(np.arange(1.1, 2.1, .1)):
    r = get_transport_reduction(u, f).mean(axis=1)
    axs.plot(u, r, '-r', linewidth=2)
    ul = 5.+1.5*i
    axs.annotate('%d%%' % ((f-1)*100), xy=(ul, np.interp(ul, u, r)), fontsize=8,
                 ha='center', va='center', backgroundcolor='w', rotation=0)

r = np.concatenate((get_transport_reduction(u, 1.1),
                    get_transport_reduction(u, 2.0)), axis=1)
xy = lines_to_polygon(u, r)
axs.add_artist(plt.Polygon(xy, alpha=.5, zorder=-1, closed=True, facecolor='none', edgecolor='none', hatch='////'))

axs.set_xlim((0, 20))
axs.set_ylim((0, 101))
axs.set_xlabel('wind velocity [$\mathrm{m/s}$]')
axs.set_ylabel('transport reduction [$\mathrm{\%}$]\n$\Delta q = q_{eq,2} / q_{eq,1} \cdot 100\%$')
axs.set_xticks(np.arange(0, 21, 5))
axs.grid()

fig.savefig('../Figures/transport_reduction.pdf')
```

# TNO

```python
fname = os.path.join(root, 'TNOdata.txt')

with open(fname, 'r') as fp:
    
    data = []
    for i, row in enumerate(fp):
        row = re.split('\t', row)
        if i == 0:
            columns = row
        else:
            for j, col in enumerate(row):
                if re.match('^[-]+$', col):
                    row[j] = np.nan
                elif re.match('^-?[0-9]+$', col):
                    row[j] = int(col)
                elif re.match('^-?[0-9\.]+$', col):
                    row[j] = float(col)
            r = [np.nan] * len(columns)
            r[:(j+1)] = row
            data.append(r)
            
    df_tno = pd.DataFrame(data, columns=columns)
```

```python
df_tno_f = df_tno.filter(regex='^F\d+').fillna(0.)#.iloc[idx]
df_tno_f.columns = [np.mean([float(x) for x in re.match('^F(\d+)-(\d+)', ix).groups()]) for ix in df_tno_f.columns]

df_tno_cs = np.cumsum(df_tno_f, axis=1)
d50 = 10**(np.interp(50, df_tno_cs.mean(axis=0), np.log10(df_tno_cs.columns)))

fig, axs = plt.subplots()
axs.plot(df_tno_cs.columns, df_tno_cs.T, '-k', alpha=.1);
axs.plot(df_tno_cs.columns, df_tno_cs.mean(axis=0), '-or');
axs.annotate('$\mathrm{d_{50} = %d}$' % d50, xy=(d50, 50),
             ha='center', va='center', bbox={'boxstyle':'round,pad=.3', 'facecolor':'w'})
axs.set_xlim((10, 1000))
axs.set_ylim((0, 100))
axs.set_yticks((0, 10, 30, 50, 60, 90, 100))
axs.set_ylabel('percentile [%]')
axs.set_xlabel('grain size [$\mathrm{\mu m}$]')
axs.set_xscale('log')
axs.grid()
```

# Imares

```python
host = 'http://opendap.tudelft.nl/thredds/dodsC/data2/zandmotor/morphology/sediment_composition/'

fnames = ['sedimentsampling_sandmotor_oct2010.nc',
          'sedimentsampling_sandmotor_aug2012.nc',
          'sedimentsampling_sandmotor_oct2013.nc']

bbox = matplotlib.path.Path(((71500, 450750),
                             (74500, 450750),
                             (74500, 453750),
                             (71500, 453750)))

p = [10, 30, 50, 60, 90]

dfs = []
for fname in fnames:
    with netcdf4_pydap.Dataset(os.path.join(host, fname)) as ds:
        t = ds.variables['type'][:]
        idx = np.asarray([ti.startswith('Strandbemonstering') for ti in t])

        x = ds.variables['x'][:][idx]
        y = ds.variables['y'][:][idx]
        
        d = np.zeros((len(p), np.sum(idx)))
        for i, pi in enumerate(p):
            d[i,:] = ds.variables['D%d' % pi][:][idx]

    idx = np.asarray(bbox.contains_points(np.asarray(zip(x,y))))
    
    year = np.asarray([int(re.sub('\D', '', fname))] * len(x))
    data = np.concatenate((year.reshape((1,-1)),
                           x.reshape((1,-1)),
                           y.reshape((1,-1)), d), axis=0).T
    
    df = pd.DataFrame(data, columns=['year', 'x', 'y'] + p).set_index(['year', 'x', 'y'])    
    dfs.append(df)
    
df_imares = pd.concat(dfs)
```
