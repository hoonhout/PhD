```python
M, legend = knmi.read_uurgeg('/Users/hoonhout/Downloads/uurgeg_330_2011-2020.txt')
```

```python
data = M['DD'].replace(990.) / 180. * np.pi
bins = np.linspace(0., 2*np.pi, 16)
ranges = [(4, 12),
          (12, 22)]

fig, axs = subplots(1, 2, figsize=(13,4), subplot_kw={'projection':'polar'})

for j, r in enumerate(ranges):
    for i, threshold in enumerate(range(r[0], r[1], 2)):
        ix = M['FH'].divide(10.) >= threshold
        if np.any(ix):
            f = np.maximum(0., .9-.05*threshold)
            axs[j].hist(data[ix], bins=bins, color=(f,f,f), edgecolor='none',
                        rwidth=.7, align='left', label='$\mathrm{u \geq %dm/s}$' % threshold)

    if j == 0:
        h = 1e4 * np.histogram(data, weights=M['FH']**3, bins=bins, normed=True)[0]
        axs[j].plot(bins, np.append(h, h[0]), '-k', linewidth=2., label='sediment\ntransport\npotential')

    axs[j].set_theta_direction(-1)
    axs[j].set_theta_offset(np.pi/2.)
    axs[j].set_rticks(())
    
items = zip(*[(o, o.get_label())
              for ax in axs
              for o in ax.get_children()
              if type(o) in [matplotlib.lines.Line2D, matplotlib.patches.Rectangle] and o.get_label() != '_nolegend_'])

axs[0].legend(*items, loc='upper left', bbox_to_anchor=(1.15,1.1))

fig.savefig('/Users/hoonhout/Checkouts/PhD/Journals/fielddata/windconditions.pdf')
```
