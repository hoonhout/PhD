```python
def dist_point2line(p, v, w):
    dist = lambda v, w: (v[0] - w[0])**2 + (v[1] - w[1])**2
    
    l2 = dist(v, w)
    if l2 == 0: return dist(p, v)
  
    t = ((p[0] - v[0]) * (w[0] - v[0]) + (p[1] - v[1]) * (w[1] - v[1])) / l2
    
#    if t < 0: return dist(p, v)
#    if t > 1: return dist(p, w)
    
    return dist(p, [v[0] + t * (w[0] - v[0]),
                    v[1] + t * (w[1] - v[1])])
```
