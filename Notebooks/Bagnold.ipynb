```python
import os
import knmi
import pytz
import netCDF4
import pandas as pd
import cPickle as pickle
from datetime import datetime, timedelta

t0 = datetime(2011, 9, 1, 0, 0, 0, 0, pytz.UTC)
t1 = datetime(2015, 9, 1, 0, 0, 0, 0, pytz.UTC)
```

```python
meteo, legend = knmi.read_uurgeg('/Users/hoonhout/Downloads/uurgeg_330_2011-2020.txt')
ix = (meteo['DD'] > 360.) | (meteo['FH'] >= 900)
meteo = meteo.ix[~ix]
u10 = meteo['FH'] / 10.
```

```python
d = 335e-6
alpha = .174/np.log10(10./.01)
uth = .085 * np.sqrt((2650.-1.25)/1.25 * 9.81 * d) / alpha
print alpha, uth, uth*alpha
```

```python
df_dir = np.maximum(0., np.cos((meteo['DD']+48.)/180.*np.pi))
df_bagnold = 1.8 * 1.25/9.81 * np.sqrt(d/250e-6) * alpha**3 * (u10 - uth)**3 # kg/m/s
df_kawamura = 2.78 * 1.25/9.81 * alpha**3 * (u10 + uth)**2 * (u10 - uth) # kg/m/s
df_horikawa = 1. * 1.25/9.81 * alpha**3 * (u10 + uth)**2 * (u10 - uth) # kg/m/s
df_lettau = 6.7 * 1.25/9.81 * np.sqrt(d/250e-6) * alpha**3 * (u10 + uth) * u10**2 # kg/m/s

def compute_volumes(df):
    #V1 = df.ix[t0:t1].sum() * 3600. * 4000 / 2650. / .6 / 1e4 # 10e4 m^3, wind energy only
    V1 = (df*(df_dir>0.)).ix[t0:t1].sum() * 3600. * 4000 / 2650. / .6 / 1e4 # 10e4 m^3, wind energy only
    V2 = (df*np.maximum(0., df_dir)).ix[t0:t1].sum() * 3600. * 4000 / 2650. / .6 / 1e4 # 10e4 m^3, cross-shore component only
    V3 = (df*np.maximum(500.,df_dir*4000.)).ix[t0:t1].sum() * 3600. / 2650. / .6 / 1e4 # 10e4 m^3, minimum orifice of 1 km
    return V1, V3, V2

for df in [df_bagnold, df_kawamura, df_horikawa, df_lettau]:
    V = compute_volumes(df)
    for Vi in V:
        print Vi, 40/Vi
    print ''
```
