```python
import os
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm

root = 'http://opendap.tudelft.nl/thredds/dodsC/data2/uuid/3bc3591b-9d9e-4600-8705-5b7eba6aa3ed/'
with netCDF4.Dataset(os.path.join(root, 'nails_135.nc'), 'r') as nc:
    t = netCDF4.num2date(nc.variables['time'][:], nc.variables['time'].units)
    x = nc.variables['x'][:]
    y = nc.variables['y'][:]
    z = nc.variables['z'][:,:]
    c = nc.variables['cross_shore'][:]
    e = nc.variables['erosion'][:,:]

n = 5#z.shape[0]

windrain = [(8.0, 0.2),
            (12.2, 0.3),
            (9.7, 0.0),
            (10.3, 0.6),
            (7.7, 0.3)]
```

```python
my_cmap = matplotlib.cm.get_cmap('OrRd')
my_cmap.set_under((.7, .7, .7))

fig, axs = plt.subplots(n, 1, figsize=(10,n*3), squeeze=True, sharex=True, sharey=True)
for i in range(n):
    #axs[i,0].scatter(x[:], y[:], c=e[i,:], s=50, cmap=my_cmap, vmin=0, vmax=10, linewidths=0)
    #axs[i,0].grid()
    #axs[i,0].set_aspect('equal')
    
    
    axs[i].fill_between(c[:], z[3,:], color=(.9,.9,.9))
    p = axs[i].scatter(c[:], z[i,:], c=e[i,:], s=50, cmap=my_cmap, vmin=0, vmax=20, linewidths=0, edgecolor='none')
    axs[i].set_title(t[i])
    axs[i].grid()
    axs[i].set_xlim((-10, 60))
    axs[i].set_ylim((1, 2.5))
    axs[i].set_yticks((1,1.5,2,2.5))
    axs[i].set_ylabel('profile height [m+NAP]')
    cb = fig.colorbar(p, ax=axs[i])
    cb.set_label('erosion [mm]')
    
    axs[i].plot([0,0], [1,2.5], '--k')
    axs[i].plot([21,21], [1,2.5], '--k')
    axs[i].annotate('mast\nlocation', xy=(0,1), xytext=(5,5), textcoords='offset points', ha='left', va='bottom')
    axs[i].annotate('mast\nlocation', xy=(21,1), xytext=(5,5), textcoords='offset points', ha='left', va='bottom')
    axs[i].annotate('mast\nlocation', xy=(60,1), xytext=(-5,5), textcoords='offset points', ha='right', va='bottom')

    axs[i].annotate('wind: %0.1f m/s\nrain: %0.1f mm' % windrain[i],
                    xy=(60,2.5), xytext=(-5,-5), textcoords='offset points', ha='right', va='top')


axs[-1].set_xlabel('cross-shore distance from most seaward mast [m]')

fig.savefig('../Figures/nails_profile.pdf')
```

```python
fig, axs = plt.subplots(n+1, 1, figsize=(10,n*3), squeeze=True, sharex=True, sharey=False)
for i in range(n):
    #axs[i].bar(c[:], e[i,:], color='k', width=.1, edgecolor='none')
    axs[i].fill_between(c[:], e[i,:], color='k')
    axs[i].set_title(t[i])
    axs[i].grid()
    axs[i].set_xlim((-10, 60))
    axs[i].set_ylim((0, 30))
    #axs[i].set_yticks((1,1.5,2,2.5))
    axs[i].set_ylabel('erosion [mm]')
    
    axs[i].annotate('wind: %0.1f m/s\nrain: %0.1f mm' % windrain[i],
                    xy=(21,30), xytext=(5,-5), textcoords='offset points', ha='left', va='top')

ix = ~np.isnan(z[3,:])
axs[-1].plot(c[ix], z[3,ix], '-k')
axs[-1].set_xlabel('cross-shore distance from most seaward mast [m]')
axs[-1].set_ylabel('profile height [m+MSL]')
axs[-1].set_ylim((1,3))
axs[-1].grid()

for ax in axs:
    m = ax.get_ylim()[1]
    ax.plot([0,0], [0,30], '--k')
    ax.plot([21,21], [0,30], '--k')
    ax.annotate('mast\nlocation', xy=(0,m), xytext=(-5,-5), textcoords='offset points', ha='right', va='top')
    ax.annotate('mast\nlocation', xy=(21,m), xytext=(-5,-5), textcoords='offset points', ha='right', va='top')
    ax.annotate('mast\nlocation', xy=(60,m), xytext=(-5,-5), textcoords='offset points', ha='right', va='top')

fig.savefig('../Figures/nails.pdf')
```
