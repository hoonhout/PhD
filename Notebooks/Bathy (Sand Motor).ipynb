```python
import os
import re
import glob
import scipy.io
import pandas as pd
import scipy.interpolate
import cPickle as pickle
from datetime import datetime

root = '/Users/hoonhout/Checkouts/Zandmotor/morphology/JETSKI/raw/zandmotor/'
```

# Raw data

```python
dx = 10.
dy = 10.

pivot = [72502.0, 452074.0]
theta = 48. / 180. * np.pi
R = np.matrix([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta),  np.cos(theta)]])

for fpath in glob.glob(os.path.join(root, '*')):
    for fname in glob.glob(os.path.join(fpath, '*.mat')):
        print fname

        # read raw data
        MAT = scipy.io.loadmat(fname)
        key = [x for x in MAT.keys() if re.match('[A-Z_]+$', x)][0] # take first key with capitals only
        
        x, y, z = tuple(MAT[key][0][0])[:3] # get first 3 columns

        # define grid
        X, Y = np.meshgrid(np.arange(70500,  75000, dx),
                           np.arange(450000, 455000, dy))

        # interpolate
        Z = scipy.interpolate.griddata((x.flatten(), y.flatten()), z.flatten(),
                                       (X.flatten(), Y.flatten())).reshape(X.shape)
        Z = ma.masked_invalid(Z)

        # rotate data
        xr, yr = tuple((np.matrix([x.flatten() - pivot[0],
                                   y.flatten() - pivot[1]]).T * R).T)
        xr = np.asarray(xr)
        yr = np.asarray(yr)
        
        # define rotated grid
        Xr, Yr = np.meshgrid(np.arange(-2000, 3000, dx),
                             np.arange(-750,  1250, dy))

        # interpolate
        Zr = scipy.interpolate.griddata((xr.flatten(), yr.flatten()), z.flatten(),
                                       (Xr.flatten(), Yr.flatten())).reshape(Xr.shape)
        Zr = ma.masked_invalid(Zr)
        
        # fill
        Zf = Zr.copy()

        for sgn in (-1,1):
            for sft in range(10):
                Zshifted = np.roll(Zf, shift=sgn*sft, axis=1)
                idx = ~Zshifted.mask * Zf.mask * (np.sign(Xr) == sgn)
                Zf[idx] = Zshifted[idx]
        for sgn in (-1,1):
            for sft in range(10):
                Zshifted = np.roll(Zf, shift=sgn*sft, axis=0)
                idx = ~Zshifted.mask * Zf.mask * (np.sign(Yr) == sgn)
                Zf[idx] = Zshifted[idx]

        # save
        with open(fname.replace('.mat', '.pkl'), 'w') as fp:
            pickle.dump({'x':x, 'y':y, 'z':z, 'X':X, 'Y':Y, 'Z':Z}, fp)
        with open(fname.replace('.mat', '_rotated.pkl'), 'w') as fp:
            pickle.dump({'x':xr, 'y':yr, 'z':z, 'X':Xr, 'Y':Yr, 'Z':Zr, 'Zf':Zf}, fp)
            
        # plot
        fig, axs = subplots(figsize=(12,4))
        p = axs.pcolormesh(Xr, Yr, Zf, cmap='gist_earth', vmin=-12, vmax=10)
        axs.plot(0, 0, 'ok')
        axs.set_aspect('equal')
        axs.set_xlim((-2000, 3000))
        axs.set_ylim((-750, 1250))
        fig.colorbar(p)
        fig.savefig(fname.replace('.mat', '.png'))
```

# Animation

```python
dts = []
z = []

for fpath in glob.glob(os.path.join(root, '*')):
    for fname in glob.glob(os.path.join(fpath, '*_rotated.pkl')):
        print fname
        
        with open(fname, 'r') as fp:
            b = pickle.load(fp)
    
        dt = datetime.strptime(os.path.split(fpath)[1], '%Y_%m_%d')
        X, Y, Z = b['X'], b['Y'], b['Zf']
        
        dts.append(dt)
        z.append(Z.flatten())
```

```python
df = pd.DataFrame(data=z, index=dts)
df = df.mask(~((df<0.) | (df>0.))).fillna(method='ffill') # remove weird values that are not smaller, not greater than zero, but also not zero or nan!
```

```python
df_hires = df.resample('1D').interpolate()
```

```python
from matplotlib import animation

fig, axs = subplots(figsize=(12,4))
axs.set_aspect('equal')
axs.set_xlim((-2000, 2750))
axs.set_ylim((-500, 1250))

def init():
    p = animate(0)
    fig.colorbar(p[0])
    return p,

def animate(i):
    Z = df_hires.iloc[i].as_matrix().reshape(X.shape)
    p = axs.pcolormesh(X, Y, Z, cmap='gist_earth', vmin=-12, vmax=10)    
    return p,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=len(df_hires), interval=100, blit=True)

# save the animation as an mp4.  This requires ffmpeg or mencoder to be
# installed.  The extra_args ensure that the x264 codec is used, so that
# the video can be embedded in html5.  You may need to adjust this for
# your system: for more information, see
# http://matplotlib.sourceforge.net/api/animation_api.html
anim.save('/Users/hoonhout/Checkouts/Zandmotor/morphology/JETSKI/raw/zandmotor/zandmotor.png',
          fps=30) #, writer='ffmpeg') #, extra_args=['-vcodec', 'libx264'])
```

```python
Z = df.iloc[0].as_matrix().reshape(X.shape)
Zmin = df.min(axis=0).as_matrix().reshape(X.shape)
Zmin[Y<-100.] = 0.

fig, axs = subplots(figsize=(12,4))
p = axs.pcolormesh(X, Y, Z[:,:-1], cmap='gist_earth', vmin=-12, vmax=10)  

# dune lake
idx = (X > -600) & (X < 100) & (Y < 0)
nx, ny = idx.sum(axis=0).max(), idx.sum(axis=1).max()
ctr = axs.contour(X[idx].reshape((nx,ny)),
                  Y[idx].reshape((nx,ny)),
                  Z[idx].reshape((nx,ny)), [1.5], alpha=0.)
axs.plot(ctr.allsegs[0][1][:,0],
         ctr.allsegs[0][1][:,1], '-r', linewidth=2)

path = matplotlib.path.Path(ctr.allsegs[0][1])
idx_dunelake = path.contains_points(np.asarray([X.flatten(),
                                                Y.flatten()]).T)

# lagoon
idx = (X > 100) & (X < 2000) & (Y < 250)
nx, ny = idx.sum(axis=0).max(), idx.sum(axis=1).max()
ctr = axs.contour(X[idx].reshape((nx,ny)),
                  Y[idx].reshape((nx,ny)),
                  Z[idx].reshape((nx,ny)), [1.5], alpha=0.)
axs.plot(ctr.allsegs[0][0][:-10,0],
         ctr.allsegs[0][0][:-10,1], '-g', linewidth=2)

path = matplotlib.path.Path(ctr.allsegs[0][0][:-10,:])
idx_lagoon = path.contains_points(np.asarray([X.flatten(),
                                              Y.flatten()]).T)

# dry beach
ctr = axs.contour(X, Y, Zmin, [2.], alpha=0.)
axs.plot(ctr.allsegs[0][0][:,0],
         ctr.allsegs[0][0][:,1], '-b', linewidth=2)

path = matplotlib.path.Path(ctr.allsegs[0][0])
idx_drybeach = path.contains_points(np.asarray([X.flatten(),
                                                Y.flatten()]).T)

fig.colorbar(p)
axs.set_aspect('equal')
axs.set_xlim((-2000, 2750))
axs.set_ylim((-500, 1250))
```
