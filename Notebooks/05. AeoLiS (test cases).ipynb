# Grainsize distribution

```python
import math
import scipy.optimize

tophi = lambda x: -np.log2(x/.001)
fromphi = lambda x: .001 * 2**-x

root2 = '../Data/grainsize'

df = pd.read_excel(os.path.join(root2, '2013 07 25 KD surface sieve data process REDO.xlsx'))
ds1 = df.ix[7::5,2:10]

df = pd.read_excel(os.path.join(root2, '2013 07 25 surface sieve data process.xlsx'))
ds2 = df.ix[52::5,2:10]

gs = np.asarray(df.ix[2,2:10] / 1000., dtype=np.float)
ds = np.asarray(pd.concat((ds1, ds2)) / 100., dtype=np.float)

phi = tophi(gs)[::-1]
dsm = np.mean(ds, axis=0)[::-1]
phim = np.interp(.5, dsm, phi)

fig, axs = plt.subplots()
axs.plot(phi, dsm, '-k')
axs.plot(phim, .5, 'ok')

fromphi(phim) * 1e6

pdf = lambda x, mu, sigma: exp(-((x-mu)/sigma)**2)
cdf = lambda x, mu, sigma: np.asarray([.5 * (1. + math.erf((xi-mu)/(sigma * np.sqrt(2)))) for xi in x])

gs2 = np.arange(50, 1180, 50) * 1e-6
ds2 = cdf(gs2, 300e-6, 200e-6)

plot(tophi(gs2), ds2, '-r')

p = scipy.optimize.curve_fit(cdf, phi, 1.-dsm)[0]
print p

phir = np.arange(-1,5,.1)
plot(phir, 1.-cdf(phir, *p), '-g')

#grain_size = np.arange(150, 1001., 50) * 1e-6
grain_size = fromphi(np.arange(0., 3., .2))
grain_dist = pdf(tophi(grain_size), *p)
grain_dist_org = pdf(grain_size, 300e-6, 200e-6)

#grain_size_c = np.asarray([2., 3., 4., 5., 10., 20.]) * 1e-3
grain_size_c = fromphi(np.arange(-5., 0., .5))
#grain_dist_c = pdf(tophi(grain_size_c), -3., .5)
grain_dist_c = pdf(tophi(grain_size_c), -2., 1.)

f = .05
grain_dist /= grain_dist.sum() / (1-f)
grain_dist_c /= grain_dist_c.sum() / f
grain_dist_org /= grain_dist_org.sum()

fig, axs = plt.subplots()
axs.plot(grain_size*1e6, grain_dist_org, '-ok')
axs.plot(grain_size*1e6, grain_dist, '-or')
axs.plot(grain_size_c*1e6, grain_dist_c, '-sr')
axs.set_xscale('log')

print fromphi(np.interp(.5, cdf(tophi(grain_size), *p), tophi(grain_size))) * 1e6
```

# Boundary conditions

```python
duration = 30 * 3600 * 24.

np.savetxt(os.path.join(root, 'wind_constant.txt'), np.asarray([[0., 12., 0.],
                                                                [duration, 12., 0.]]))

np.savetxt(os.path.join(root, 'wind.txt'), np.asarray([[0., 12., 0.],
                                                       [7*24*3600.-1, 12., 0.],
                                                       [7*24*3600., 30., 0.],
                                                       [8*24*3600.-1, 30., 0.],
                                                       [8*24*3600., 12., 0.],
                                                       [15*24*3600.-1, 12., 0.],
                                                       [15*24*3600., 30., 0.],
                                                       [16*24*3600.-1, 30., 0.],
                                                       [16*24*3600., 12., 0.],
                                                       [duration, 12., 0.]]))

np.savetxt(os.path.join(root, 'tide_constant.txt'), np.asarray([[0., 0.],
                                                                [duration, 0.]]))

with open(os.path.join(root, 'tide.txt'), 'w') as fp:
    for i in range(0, int(duration)+1, 600):
        m = np.sin(i/86400.0*4.0*pi) + 0.0 * np.sin(i/30.0/86400.0*4.0*pi)
        fp.write('%10.2f %10.2f\n' % (i, m))
        
np.savetxt(os.path.join(root, 'waves.txt'), np.asarray([[0., 1.],
                                                        [duration, 1.]]))
```

```python
from scipy.optimize import curve_fit

fig, axs = plt.subplots(1, 2, figsize=(10,4), gridspec_kw={'width_ratios':[3,1]})

d = -np.asarray((data['case3b']-data['case1b'])/data['case1b'])
n = len(d)
s = axs[0].scatter(uw[:n], 100. * d, c=np.arange(len(d))/24.,
                   cmap='Greys', vmin=0, vmax=30, marker='o', alpha=.5, edgecolor='k')

df = pd.DataFrame({'u':uw[:n], 'd':d})
df = df.groupby(by=np.digitize(df['u'], np.arange(0., 30., .5)))
df = df.median()

i = 12
#fcn = lambda x, a, mu, sigma: a + (1.-a) / (1. + exp(-(mu-x)/sigma))
fcn = lambda x, a, b: np.minimum(1., exp(-a*(x-b)))
p, pcov = curve_fit(fcn, df['u'].ix[i:], df['d'].ix[i:], p0=(.1, 7.))
R2 = 1. - np.sum((df['d'] - fcn(df['u'], *p))**2) / np.sum((df['d'] - df['d'].mean())**2)
formula = '$\\exp \\left( -%0.2f (u - %0.2f) \\right)$' % tuple(p)
axs[0].plot(df['u'], 100. * fcn(df['u'], *p), '-r', linewidth=2, label='curve fit ($R^2 = %0.2f$)' % R2)
axs[0].plot(df['u'], 100. * df['d'], '^r', label='median reduction')
axs[0].legend(loc='lower left')
axs[0].annotate(formula, xy=(10,100.*fcn(10, *p)), xytext=(5,30),
                fontsize=16, color='r', ha='center', arrowprops={'arrowstyle':'-','color':'r'}, bbox={'color':'w'})
print p

#Rmin = np.nanmin(R[:,0])
#umin = u[R[:,0] == Rmin]
#axs[0].annotate('min. %0.1f%%' % Rmin, xy=(umin, Rmin), xytext=(8, -75),
#                arrowprops=dict(arrowstyle='-|>', linewidth=2),
#                color='k', fontsize=12, fontweight='bold', ha='center', va='center',
#                bbox=dict(boxstyle='round, pad=.2', fc='w', ec='k'))

axs[0].set_xlabel('hourly averaged wind velocity [m/s]')
axs[0].set_ylabel('average relative reduction in sediment transport\nin case P3b with respect to case P1b [%]')
axs[0].set_xlim((0, 25))
axs[0].set_ylim((0,100))
axs[0].legend(loc='lower left')
axs[0].grid()
#axs[0].set_xticks([6.89])

axs[1].hist(uw0[:,1], color='k', weights=[1/60.]*len(uw0), bins=np.arange(0,30))
axs[1].set_xlim((0., 25.))
axs[1].set_xlabel('wind velocity [m/s]')
axs[1].set_ylabel('ocurrence [h]')
axs[1].yaxis.set_ticks_position('right')
axs[1].yaxis.set_label_position('right')
axs[1].grid()

cb = fig.colorbar(s, ax=axs[0])
cb.set_label('time [days]')
cb.set_ticks(np.arange(0,31,5))

fig.savefig('../Figures/variable_wind_py.pdf')

print fcn(12, *p)
df.ix[17]
```
