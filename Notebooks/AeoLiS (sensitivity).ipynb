```python
import aeolis.io

tophi = lambda x: -np.log2(x/.001)
fromphi = lambda x: .001 * 2**-x
pdf = lambda x, mu, sigma: exp(-((x-mu)/sigma)**2)

p = [1.57749633, 0.42538961]

grain_size = fromphi(np.arange(0., 3., .2))
grain_dist = pdf(tophi(grain_size), *p)

n = len(grain_size)
```

```python
pars_case1 = aeolis.io.read_configfile('/Users/hoonhout/GitHub/aeolis-models/jgr/testcases/case1.txt', load_defaults=False)
pars_case3 = aeolis.io.read_configfile('/Users/hoonhout/GitHub/aeolis-models/jgr/testcases/case3.txt', load_defaults=False)

pars_case1.update({'dt':600., 'output_times':86400.})
pars_case3.update({'dt':600., 'output_times':86400.})
```

```python
def strip_shells(pars):
    pars2 = deepcopy(pars)
    pars2['grain_size'] = pars['grain_size'][:15]
    pars2['grain_dist'] = pars['grain_dist'][:15]
    pars2['nfractions'] = 15
    return pars2
```

```python
import glob
import os

root = '/Users/hoonhout/GitHub/aeolis-models/jgr/sensitivity/'

def get_transportrate(fltr, par='par'):
    p = []
    data = []
    for fname in glob.glob(os.path.join(root, fltr)):
        with xray.open_dataset(fname) as ds:
            if ds['Ct.avg'].shape[0] == 0:
                continue
            data.append(ds['Ct.avg'][:,0,-1,:].sum(axis=0).sum(axis=-1).copy())
        p.append(float(re.search(fltr.replace('*', '([-\d\.]+)'), fname).groups()[0]))
    data = xray.concat(data, dim=par)
    data[par] = p
    
    return data.to_dataframe().unstack()

fig, axs = subplots(2, 3, figsize=(12,8))
axs = axs.flatten()

df = get_transportrate('case3_bi*.nc', par='bi')
df['Ct.avg'] = df['Ct.avg'].subtract(df['Ct.avg'][0]).divide(df['Ct.avg'][0])
df['Ct.avg'].T.plot(legend=False, color='k', ax=axs[0], linewidth=2)
#axs[0].plot(.2, df['Ct.avg'].iloc[4], 'ok')
#axs[0].xaxis.set_label_position('top')
#axs[0].xaxis.set_ticks_position('top')
#axs[0].yaxis.set_label_position('right')
#axs[0].yaxis.set_ticks_position('right')
axs[0].set_xlabel(r'bed interaction parameter [-]', labelpad=16)
#axs[0].set_ylim((-1., 1.))

df = get_transportrate('case3_dod*.nc', par='facDOD')
df['Ct.avg'] = df['Ct.avg'].subtract(df['Ct.avg'][.1]).divide(df['Ct.avg'][.1])
df['Ct.avg'].T.plot(legend=False, color='k', ax=axs[1], linewidth=2)
#axs[1].plot(.1, df['Ct.avg'].iloc[2], 'ok')
axs[1].set_xlabel(r'depth of disturbance factor [-]', labelpad=16)
axs[1].set_xlim((0., .5))
#axs[1].set_ylim((-1., 1.))

df = get_transportrate('case3_T*.nc', par='Tdry')
df['Ct.avg'] = df['Ct.avg'].subtract(df['Ct.avg'][3]).divide(df['Ct.avg'][3])
df['Ct.avg'].T.plot(legend=False, color='k', ax=axs[2], linewidth=2)
#axs[2].plot(6., df['Ct.avg'].iloc[2], 'ok')
axs[2].set_xlabel('drying time scale [h]', labelpad=16)
#axs[2].yaxis.set_label_position('right')
#axs[2].yaxis.set_ticks_position('right')
#axs[2].set_xscale('log')
#axs[2].set_ylim((-1., 1.))
axs[2].set_xticks([.1,1.,2.,3.,4.,5.,6.])

df = get_transportrate('case3_s*.nc', par='s')
df['Ct.avg'] = df['Ct.avg'].subtract(df['Ct.avg'][.4]).divide(df['Ct.avg'][.4])
df['Ct.avg'].T.plot(legend=False, color='k', ax=axs[3], linewidth=2)
#axs[3].plot(.4, df['Ct.avg'].iloc[3], 'ok')
#axs[3].xaxis.set_label_position('top')
#axs[3].xaxis.set_ticks_position('top')
axs[3].set_xlabel('grain size distribution\nstandard deviation [$\Phi$]', labelpad=-4.)
axs[3].set_xticks([0, .5, 1, 2, 3])
axs[3].set_xticklabels(('single\nfraction', .5, 1., 2., 'uniform'))
axs[3].set_xlim((0,3))
#axs[3].set_ylim((-1., 1.))

for ax in axs[:3]:
    ax.set_ylim((-.1, .1))
axs[3].set_ylim((-1., 1.))

axs[0].set_ylabel('normalized sediment transport rate [-]')
axs[3].set_ylabel('normalized sediment transport rate [-]')
axs[1].set_yticklabels(())
axs[2].set_yticklabels(())

df = get_transportrate('case3_u*.nc', par='uw')
df2 = get_transportrate('case1_u*.nc', par='uw')
df2['Ct.avg'][4] = np.nan

dfref = df['Ct.avg'].divide(df2['Ct.avg'])
dfref[4] = 0.

dfref.T.plot(legend=False, color='k', ax=axs[4], linewidth=2)
#axs[4].plot(8., 1., 'ok')

#df['Ct.avg'] = df['Ct.avg'].divide(ref)
#df['Ct.avg'].T.plot(legend=False, color='k', ax=axs[4], linewidth=2, linestyle='--')
axs[4].set_xlabel(r'wind velocity [m/s]', labelpad=16)
axs[4].set_xticks(range(4,25,2))
axs[4].set_xlim((4., 20.))
axs[4].set_ylim((0., .4))
axs[4].set_ylabel('sediment transport rate\nrelative to case #1 [-]')

axs[5].set_axis_off()
#axs[2].tick_params(axis='x', colors='red')
#axs[3].tick_params(axis='y', colors='red')
#axs[4].tick_params(axis='y', colors='red')

for i, ax in enumerate(axs[:3]):
    ax.annotate('%s)' % chr(97+i), xy=(.95,.95), xycoords='axes fraction',
                fontsize=20, fontweight='bold', color='k',
                ha='right', va='top')
for i, ax in enumerate(axs[3:-1]):
    ax.annotate('%s)' % chr(97+i+3), xy=(.95,.05), xycoords='axes fraction',
                fontsize=20, fontweight='bold', color='k',
                ha='right', va='bottom')
    
for i, ax in enumerate(axs[3:-1]):
    pos = ax.get_position()
    pos.x0 += -.05*(1-i)+1./6
    pos.x1 += -.05*(1-i)+1./6
    pos.y0 -= .05
    pos.y1 -= .05
    ax.set_position(pos)

fig.savefig('/Users/hoonhout/Checkouts/PhD/Thesis2/sensitivity_py.pdf')
```
