```python
import numpy as np
import cPickle as pickle
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
with open('../Data/budgets/sourcessinks.pkl', 'r') as fp:
    sinks = dict(pickle.load(fp)[1])
    
with open('../Data/lidar/lidar2.pkl', 'r') as fp:
    lidar = pickle.load(fp)
    xi = lidar['X']
    yi = lidar['Y']
```

```python
def running_mean(x, N):
    x[np.isnan(x)] = 0.
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    X = np.nan + np.zeros(x.shape)
    X[(N/2-1):-(N/2)] = (cumsum[N:] - cumsum[:-N]) / N
    return X


def rotate(x, y, phi, origin=(72502.0, 452074.0)):

    x = np.asarray(x)
    y = np.asarray(y)
    phi = phi / 180. * np.pi

    R = np.asarray([[np.cos(phi), -np.sin(phi)],
                    [np.sin(phi),  np.cos(phi)]])

    XY = np.concatenate((x.reshape(-1,1) - origin[0],
                         y.reshape(-1,1) - origin[1]), axis=1)

    XY = np.dot(XY, R)
    xr = XY[:,0].reshape(x.shape)
    yr = XY[:,1].reshape(y.shape)

    return xr, yr


def draw_northarrow(axs, xy, scale, alpha):    
    verts = np.asarray([[ 0., -.5],
                        [-.7, -1.],
                        [ 0.,  1.],
                        [ .7, -1.],
                        [-.15, 1.5]]).T
    
    x, y = rotate(verts[0,:], 
                  verts[1,:], 
                  alpha, 
                  origin=(0., 0.))
    
    verts = np.concatenate((xy[0]+scale*x, xy[1]+scale*y)).reshape(verts.shape).T
        
    axs.add_patch(plt.Polygon(verts[:-1,:], facecolor='k', edgecolor='k'))
    axs.annotate('N', xy=verts[-1,:], ha='center', va='center', weight='bold',
                 fontsize=scale/30., rotation=-alpha, zorder=20)
```

```python
d = 1
ref = 3.
porfac = (1. - 0.365) / .6 # porosity
dt = (datetime(2015,3,10) - datetime(2012,1,31)).days / 365.2425

fig, axs = plt.subplots(2, 1, figsize=(10, 4), sharex=True, gridspec_kw={'height_ratios':[1,1]})
axs[0].pcolormesh(xi[::d,::d], yi[::d,::d], lidar['Z']['20120131'][::d,::d], cmap='Greys', vmin=0, vmax=15, rasterized=True)

ix = yi[:,0] < -400
V1 = np.nansum(100. * np.maximum(0., lidar['Z']['20120131'][ix,:] - ref), axis=0) / 10. / dt * porfac
V2 = np.nansum(100. * np.maximum(0., lidar['Z']['20150310'][ix,:] - ref), axis=0) / 10. / dt * porfac
dV = running_mean(V2-V1, 20)

dV[(xi[0,:]>6500)&(xi[0,:]<10000)] = np.nan

#axs[1].annotate('(ignored)', xy=(8250, 12.5), va='center', ha='center')
axs[0].annotate('$\leftarrow$ Rotterdam harbour', xy=(-5500, 800), va='bottom', ha='left')
axs[0].annotate('Scheveningen harbour $\\rightarrow$', xy=(6500, 800), va='bottom', ha='right')

draw_northarrow(axs[0], (6000, 400), 200, 48.)

axs[0].set_ylabel('cross-shore\ndistance [m]')
axs[0].set_aspect('equal', adjustable='box-forced')
#axs[0].set_xlim((-9000, 14000))
axs[0].set_xlim((-5500, 6500))
axs[0].set_ylim((-1200, 800))
axs[0].set_yticks((-1200, 0, 800))
axs[0].grid()

axs[1].fill_between(xi[0,:], dV, facecolor='r', edgecolor='none')

ix = (xi[0,:]>=-1500)&(xi[0,:]<=2500)
dVm = np.mean(dV[ix])
dVe = (sinks['lagoon'] + sinks['dune lake'])*1e4/np.sum(ix)/10./4.
axs[1].fill_between(xi[0,ix], dVm + dVe + np.zeros(xi[0,ix].shape),
                    facecolor='r', edgecolor='none', alpha=.25)
axs[1].annotate('volume accumulated\nin dune lake and lagoon', xy=(500,25), xytext=(-3000, 40), 
                arrowprops={'arrowstyle':'->'}, ha='center', va='center')

ix = (xi[0,:]>=-5500)&(xi[0,:]<=-1500)
dVm1 = np.mean(dV[ix])

ix = (xi[0,:]>=2500)&(xi[0,:]<=6500)
dVm3 = np.mean(dV[ix])
dVm2 = dVm + dVe

axs[1].plot([-5500, -1500], [dVm1, dVm1], '--k')
#axs[1].plot([-1500, 2500], [dVm2, dVm2], '--k')
axs[1].plot([-1500, 2500], [dVm, dVm], '--k')
axs[1].plot([2500, 6500], [dVm3, dVm3], '--k')

axs[1].annotate('%0.0f' % dVm1, xy=(-5500, dVm1), va='bottom', ha='left')
axs[1].annotate('%0.0f' % dVm, xy=(500, dVm), va='bottom', ha='center')
axs[1].annotate('%0.0f' % dVm3, xy=(6500, dVm3), va='bottom', ha='right')

axs[1].set_ylim((0,50))
axs[1].set_yticks((0, 25, 50))
axs[1].set_xlabel('alongshore distance [m]')
axs[1].set_ylabel('dune growth rate\n[$\mathrm{m^3/m/yr}$]')
axs[1].grid()

plt.draw()

axf = fig.add_axes([0,0,1,1], axis_bgcolor='none')
p1 = axs[0].transData.transform((2500, 800))
p2 = axs[1].transData.transform((-1500, 0))
p1 = axf.transData.inverted().transform(p1)
p2 = axf.transData.inverted().transform(p2)
axf.add_patch(plt.Rectangle(p2, p1[0]-p2[0], p1[1]-p2[1], facecolor='none', edgecolor='k', linewidth=2))
axf.annotate('Sand Motor region', xy=(p2[0], p1[1]), va='bottom', ha='left')
axf.set_axis_off()

fig.savefig('../Figures/adjacentcoasts.pdf')

print dVm1, dVm2, dVm3, dVm, np.mean((dVm2 / dVm1, dVm2 / dVm3))
```
