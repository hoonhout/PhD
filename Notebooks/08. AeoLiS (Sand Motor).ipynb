# Model setup

```python
def rotate(x, y, phi, origin=(72502.0, 452074.0)):

    x = np.asarray(x)
    y = np.asarray(y)
    phi = phi / 180. * np.pi

    R = np.asarray([[np.cos(phi), -np.sin(phi)],
                    [np.sin(phi),  np.cos(phi)]])

    XY = np.concatenate((x.reshape(-1,1) - origin[0],
                         y.reshape(-1,1) - origin[1]), axis=1)

    XY = np.dot(XY, R)
    xr = XY[:,0].reshape(x.shape)
    yr = XY[:,1].reshape(y.shape)
    
    return xr, yr


# select and rotate bathy
alpha = 48.
X, Y = rotate(grid['X'], grid['Y'], alpha)
Z = grid['Z'][key]
```

## Bathy

```python
# interpolate bathy to computational grid and write to file

if not os.path.exists(os.path.join(root, 'bathy')):
    os.mkdir(os.path.join(root, 'bathy'))
    
d = 50
x = np.arange(-500,  1000, d)
y = np.arange(-1500, 2500, d)

Xr, Yr = np.meshgrid(x, y)
np.savetxt(os.path.join(root, 'x.txt'), Xr)
np.savetxt(os.path.join(root, 'y.txt'), Yr)

for i, key in enumerate(sorted(grid['Z'].keys())):
    Z = grid['Z'][key]
    
    ix = ~np.isnan(Z.flatten())
    Zr = scipy.interpolate.griddata((Y.flatten()[ix],
                                     X.flatten()[ix]),
                                    Z.flatten()[ix],
                                    (Xr.flatten(),
                                     Yr.flatten()), method='nearest').reshape(Xr.shape)

    np.savetxt(os.path.join(root, 'bathy', 'z_%s.txt' % key), Zr)

print Zr.shape
```

## Masks

```python
# compute tidal masks
for fname in glob.glob(os.path.join(root, 'bathy', 'z_*.txt')):
    
    Z = np.loadtxt(fname)
    
    # get contour
    yi, xi = np.where(Z > 3.)
    ix = pd.DataFrame({'x':xi,'y':yi}).groupby('y').max().values.flatten()

    # convert to mask
    mask1 = 10+Xr//50 < ix.reshape((-1,1)).repeat(30,axis=1)
    mask1 = 1. - mask1
    
    # get tangent
    ix[45:] = np.interp(np.arange(45,80),[45,79],ix[[45,79]])

    # convert to mask
    mask2 = 10+Xr//50 < ix.reshape((-1,1)).repeat(30,axis=1)
    mask2 = 1. - mask2
    
    # interpolate transition
    mask = mask1.copy()
    ix = (mask1 == mask2)
    F = scipy.interpolate.LinearNDInterpolator(np.concatenate((Xr[ix].reshape((-1,1)),
                                                               Yr[ix].reshape((-1,1))), axis=1),
                                               mask[ix].reshape((-1,1)))
    mask[~ix] = F(Xr[~ix], Yr[~ix]).flatten()
    
    # add complex component
    mask = np.asarray(mask, dtype=np.complex)
    mask += (1. - mask) * 1j
    
    np.savetxt(fname.replace('z_', 'mask_'), mask)
```

```python
def get_params(sigma, Tdry, T):
    
    p = dict(nx = Zr.shape[1]-1,
             ny = Zr.shape[0]-1,
             xgrid_file = 'x.txt',
             ygrid_file = 'y.txt',
             bed_file = 'bathy/z_%s.txt' % datetime.fromordinal(o[0]).strftime('%Y%m%d'),
             wind_file = 'wind.txt',
             tide_file = 'tide.txt',
             wave_file = 'waves.txt',
             #meteo_file = 'meteo.txt',
             mask_file = 'bathy/mask_%s.txt' % datetime.fromordinal(o[0]).strftime('%Y%m%d'),
             grain_size = [0.000177, 0.000250, 0.000354, 0.000500, 0.000707, 0.001000, 0.002000, 0.004000, 0.008000, 0.016000, 0.032000],
             grain_dist = [0.005709, 0.234708, 0.608887, 0.099666, 0.001029, 0.000001, 0.010486, 0.028503, 0.010486, 0.000522, 0.000004],
             nfractions = 11,
             nlayers = 10,
             tstop = (t1-t0).days * 24. * 3600.,
             output_times = 3600 * 24. * 7.,
             output_types = [],
             output_vars = ['zb', 'Ct', 'Ct.avg', 'Cu', 'Cu.avg', 'uth', 'moist', 'uw.avg', 'mass', 'pickup', 'pickup.sum'],
             dt = 3600.,
             bi = .05,
             beta = 130.,
             sigma = float(sigma),
             Tdry = Tdry * 3600.,
             T = T,
             #Tsalt = Tsalt * 3600.,
             #th_salt = True,
             scheme = 'euler_backward',
             bedupdate = False,
             runup = True,
             callback = 'callback.py:update_bed')
    
    return p
```

```python
with open(os.path.join(root, 'callback.py'), 'w') as fp:
    fp.write('''

import os
import logging
import numpy as np
from datetime import datetime, timedelta

current = None
t0 = datetime(2011,8,1)
    
# initialize logger
logger = logging.getLogger(__name__)


def update_bed(model):
    global current
    
    k = (t0 + timedelta(seconds=model.get_current_time())).strftime('%Y%m%d')
    if current is None or current != k:
        if update_var(model, 'zb', 'bathy/z_%s.txt' % k):
            logger.info('Updated bathymetry to "%s"' % k)
            current = k
        if update_var(model, 'mask', 'bathy/mask_%s.txt' % k, dtype=np.complex):
            logger.info('Updated mask to "%s"' % k)
            

def update_var(model, var, fname, dtype=np.float):
    if os.path.exists(fname):
        try:
            val = model.get_var(var, clear=False)
            val_new = val.copy()
            val_new[:,:] = np.loadtxt(fname, dtype=dtype)
            model.set_var(var, val_new)
            return True
        except:
            return False
    else:
        return False
        
''')
```
