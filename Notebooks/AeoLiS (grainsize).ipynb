```python
def get_transport_reduction(u, f,
                            A=100., rho_p=2650., rho_a=1.25, g=9.81,
                            d_n=np.arange(200e-6, 455e-6, 10e-6)):
    
    f = sqrt(f)
    d_n, u = np.meshgrid(d_n, u)

    uth_1 = A * np.sqrt((rho_p-rho_a)/rho_p*g*d_n)
    uth_2 = A * np.sqrt((rho_p-rho_a)/rho_p*g*f*d_n)

    q_1 = np.maximum(0., u - uth_1)**3
    q_2 = np.maximum(0., u - uth_2)**3

    r = (1 - q_2 / q_1) * 100.
    r[np.isnan(r)] = 100.
    
    return r


def lines_to_polygon(u, r):

    x = np.concatenate((u[:], u[-1:0:-1]))
    y = np.concatenate((np.min(r, axis=1),
                        np.max(r, axis=1)[-1:0:-1]))
    xy = np.concatenate((x[:,np.newaxis], y[:,np.newaxis]), axis=1)
    
    return xy


u = np.arange(0., 41., .1)

fig, axs = subplots()

for i, f in enumerate(np.arange(1.1, 2.1, .1)):
    r = get_transport_reduction(u, f).mean(axis=1)
    axs.plot(u, r, '-r', linewidth=2)
    ul = 5.+1.5*i
    axs.annotate('%d%%' % ((f-1)*100), xy=(ul, np.interp(ul, u, r)), fontsize=8,
                 ha='center', va='center', backgroundcolor='w', rotation=0)

r = np.concatenate((get_transport_reduction(u, 1.1),
                    get_transport_reduction(u, 2.0)), axis=1)
xy = lines_to_polygon(u, r)
axs.add_artist(plt.Polygon(xy, alpha=.5, zorder=-1, closed=True, facecolor='none', edgecolor='none', hatch='////'))

axs.set_xlim((0, 20))
axs.set_ylim((0, 101))
axs.set_xlabel('wind velocity [$\mathrm{m/s}$]')
axs.set_ylabel('transport reduction [$\mathrm{\%}$]\n$\Delta q = q_{eq,2} / q_{eq,1} \cdot 100\%$')
axs.set_xticks(np.arange(0, 21, 5))
axs.grid()

fig.savefig('/Users/hoonhout/Checkouts/PhD/Journals/grainsize/transport_reduction.pdf')
```

```python
def get_distribution(xi=1., T=10., nf=1):

    Cu = 1.
    Ct = 0.
    Se = 1.

    p = []
    for i in range(100):

        p_air = Ct / Cu 
        p_bed = Se / sum(Se)

        p.append((Ct, (1 - xi) * p_air, p_bed * (1 - min(1, (1 - xi) * sum(p_air)))))

        Ct = Ct + (Cu - Ct) / T

    return np.asarray(p)
```

```python
def plot_awesome_arrows(x, y, f1, f2, color='k', ax=None):
    if not ax:
        fig, ax = subplots()
    else:
        fig = ax.get_figure()
        
    ax.annotate('', xy=(x, y), xytext=(x, f2 * y),
                arrowprops=dict(arrowstyle='<|-|>', color=color, linewidth=2))
    ax.annotate('', xy=(x, f2 * y), xytext=(x, .0),
                arrowprops=dict(arrowstyle='<|-|>', color=color, linewidth=2))
    ax.annotate('%d%%' % round(f1*100), xy=(x, f2+(f1-f2)/2 * y), xytext=(5,0),
                textcoords='offset points', color=color, va='center')
    ax.annotate('%d%%' % round(f2*100), xy=(x, f2/2 * y), xytext=(5,0),
                textcoords='offset points', color=color, va='center')
    
    return fig, ax


fig, axs = subplots(2, 2, figsize=(10,4))

xis = [0., .2, .5, 1.]
for i, ax in enumerate(axs.flatten()):
    
    A = get_distribution(xi=xis[i])
    x = np.arange(A.shape[0])
    
    ax.plot(A[:,0], '-k')
    ax.plot(A[:,0] * A[:,2], '-k')
    ax.fill_between(x, A[:,0] * A[:,2],
                    color=(1.,1.,.7))
    ax.fill_between(x, A[:,0] * A[:,2], A[:,0],
                    color=(.8,.8,1.))
    ax.set_xticks(())
    ax.set_xlim((0, 75))
    ax.grid()
    
    if i == 0:
        x70 = np.interp(.7, A[:,0], x)
        plot_awesome_arrows(x70, .7, .7, .3, color='k', ax=ax)
    if i == 1:
        x70 = np.interp(.7, A[:,0], x)
        plot_awesome_arrows(x70, .7, .7*.8, 1-.7*.8, color='k', ax=ax)
    if i > 1:
        ax.set_xlabel('distance downwind $\\rightarrow$')
    if mod(i,2) == 0:
        ax.set_ylabel('saturation $C_u / C_t$')
    if i == 3:
        p1 = plt.Rectangle((0,0), 0, 0, facecolor=(1.,1.,.7))
        p2 = plt.Rectangle((0,0), 0, 0, facecolor=(.8,.8,1.))
        ax.legend((p1, p2), ('contribution from bed', 'contribution from air'), loc='lower right')

    ax.set_title('$\\xi = %0.1f$' % xis[i])
    
fig.savefig('/Users/hoonhout/Checkouts/PhD/Journals/grainsize/bed_interaction_parameter.pdf')
```
