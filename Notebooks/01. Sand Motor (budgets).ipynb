```python
import os
import glob
import scipy.io
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.path
import matplotlib.patches
import matplotlib.dates
import matplotlib.collections
import matplotlib.cm
import matplotlib._cntr as cntr
import scipy.interpolate
import scipy.spatial
import cPickle as pickle
from waterbase import Waterbase
import knmi

root = '../Data/jetski'

t0 = datetime(2011, 9, 1)
t1 = datetime(2015, 9, 1)
```

# read trajectories

```python
dfs = []
for fpath in os.listdir(root):
    if os.path.isdir(os.path.join(root, fpath)) and not fpath.startswith('.'):

        date = fpath.replace('_','')
        for fname in glob.glob(os.path.join(root, fpath, '*.mat')):            
            print fname
            
            if '2016' in fname:
                continue

            data = scipy.io.loadmat(fname)
            keys = [k for k in data.keys() if not k.startswith('_')]
            
            key = 'XYZ'
            if key not in keys:    
                key = keys[0]
                
            print '    using key "%s" from "%s"' % (key, keys)
            
            X, Y, Z = \
                np.asarray(data[key][0][0][0]).flatten(), \
                np.asarray(data[key][0][0][1]).flatten(), \
                np.asarray(data[key][0][0][2]).flatten()
            d = datetime.strptime(date, '%Y%m%d')
            
            idx = pd.MultiIndex.from_tuples(zip([d] * len(Z), range(len(Z))), names=('date', 'nr'))
            df = pd.DataFrame({'X':X, 'Y':Y, 'Z':Z}, index=idx)
            
            dfs.append(df)

df_trajects = pd.concat(dfs)
```

# create grid

```python
d = 10

x0, y0 = df_trajects.min()[:2]
x1, y1 = df_trajects.max()[:2]

x0, y0 = 70000, 450000
x1, y1 = 75000, 455000

grid = {'X':[], 'Y':[], 'Z':{}}
grid['X'], grid['Y'] = np.meshgrid(np.arange(x0, x1, d),
                       np.arange(y0, y1, d))

for date in df_trajects.index.levels[0]:
    
    print date
    
    traject = df_trajects.xs(date, level=0)
    idx = (traject['X'] >= x0) & \
          (traject['X'] <= x1) & \
          (traject['Y'] >= y0) & \
          (traject['Y'] <= y1)
    
    grid['Z'][date.strftime('%Y%m%d')] = \
        scipy.interpolate.griddata((traject['X'][idx],
                                    traject['Y'][idx]),
                                                 traject['Z'][idx],
                                                 (grid['X'].flatten(),
                                                  grid['Y'].flatten())).reshape(grid['X'].shape)
        
with open(os.path.join(root, 'bathy_gridded.pkl'), 'w') as fp:
    pickle.dump(grid, fp)
```

# domain decomposition

```python
refx = np.asarray([-1500, -1000, 0, 800, 1100, 2500])
refy = np.asarray([-600, -500, -250, 200, 800])

def isiterable(x):
    try:
        _ = [i for i in x]
    except:
        return False
    return True
    

def compute_convexhull(paths, key):
    verts = np.concatenate([p[key].vertices
                            for k, p in paths.iteritems()
                            if not k.startswith('_')], axis=0)
    chull = scipy.spatial.ConvexHull(verts)

    fig, axs = plt.subplots()
    axs.plot(verts[:,0], verts[:,1], '-k')
    axs.plot(chull.points[chull.vertices,0], chull.points[chull.vertices,1], '-r')
    
    return XPath(chull.points[chull.vertices,:])


def rotate(x, y, phi, origin=(72502.0, 452074.0)):

    x = np.asarray(x)
    y = np.asarray(y)
    phi = phi / 180. * np.pi

    R = np.asarray([[np.cos(phi), -np.sin(phi)],
                    [np.sin(phi),  np.cos(phi)]])

    XY = np.concatenate((x.reshape(-1,1) - origin[0],
                         y.reshape(-1,1) - origin[1]), axis=1)

    XY = np.dot(XY, R)
    xr = XY[:,0].reshape(x.shape)
    yr = XY[:,1].reshape(y.shape)
    
    return xr, yr


class XPath(matplotlib.path.Path):
    
    
    def __init__(self, vertices, codes=None, **kwargs):
        vertices = self._vertices(vertices)
        
        super(XPath, self).__init__(vertices, codes=codes, **kwargs)
        
        self.reset_codes().close()
        
        
    def __copy__(self):
        return XPath(self.vertices, codes=self.codes)
    
    
    def __getitem__(self, s):
        return self.copy().slice(s)
    
    
    def plot(self, ax=None, **kwargs):

        if ax is None:
            fig, ax = plt.subplots()
        else:
            fig = ax.get_figure()
            
        verts = self.vertices
        verts[-1] = verts[0]
        
        ax.plot(verts[:,0], verts[:,1], **kwargs)
        return fig, ax
    
        
    def rotate(self, phi, origin=(0.,0.)):
        self.vertices = zip(*rotate(*zip(*self.vertices), phi=phi, origin=origin))
        return self

    
    def reverse(self):
        self.vertices = self.vertices[-2::-1,:]
        self.codes[1:-2] = self.codes[-2:0:-1]
        return self
    
    
    def get_distances(self):
        return np.sqrt((self.vertices[1:,0] - self.vertices[:-1,0])**2 + \
                       (self.vertices[1:,1] - self.vertices[:-1,1])**2)

    
    def close(self):
        if self.codes is not None:
            if self.codes[-1] != matplotlib.path.Path.CLOSEPOLY:
                self.vertices = np.append(self.vertices, [[np.nan,np.nan]], axis=0)
                self.codes = self.codes + [matplotlib.path.Path.CLOSEPOLY]
        return self


    def open(self):
        if self.codes is not None:
            if self.codes[-1] == matplotlib.path.Path.CLOSEPOLY:
                self.pop()
        return self
    

    def pop(self, n=1):
        self.slice(slice(0,-n))
        return self
    
    
    def slice(self, s):
        self.vertices = self.vertices[s,:]
        if self.codes is not None:
            self.codes = list(np.asarray(self.codes)[s])
            self.codes[0] = matplotlib.path.Path.MOVETO
        return self
    
    
    def split(self, *ixs):
        parts = []
        i = 0
        for ix in list(ixs) + [None]:
            parts.append(self.copy().slice(slice(i, ix)).reset_codes())
            i = ix
        return parts

        
    def clip_to_polygon(self, polygon, inverse=False, replace=False):
        poly = self._path(polygon)
        if inverse:
            idx = ~poly.contains_points(self.vertices)
        else:
            idx =  poly.contains_points(self.vertices)
        if replace:
            i = np.where(idx)[0]
            if len(i) > 0:
                pth1, pth2 = self.split(i[0])
                return pth1.extend(poly).extend(pth2).close()
        else:
            self.slice(idx).close()
        return self
    
    
    def reset_codes(self):
        self.open()
        n = self.vertices.shape[0]
        self.codes = [matplotlib.path.Path.MOVETO] + \
                     [matplotlib.path.Path.LINETO] * (n-1)
        self.close()
        return self


    def extend(self, path, merge=True):
        
        path = self._path(path)

        if merge:
            path.open()
            self.open()

        codes = list(path.codes)
        vertices = np.asarray(path.vertices)
            
        #print self.vertices.shape, vertices.shape
        self.vertices = np.concatenate((self.vertices, vertices), axis=0)
        
        if merge:
            self.reset_codes()
        else:
            self.codes = list(self.codes) + codes
            
        return self
    
    
    def contains_points(self, points, **kwargs):
        ixs = np.where(np.asarray(self.codes)[1:] == matplotlib.path.Path.MOVETO)[0] + 1
        parts = self.split(*ixs)
        for i, part in enumerate(parts):
            if i == 0:
                contains = super(XPath, part).contains_points(points, **kwargs)
            else:
                ix = super(XPath, part).contains_points(points, **kwargs)
                contains[ix] = ~contains[ix] # negate matches
        return contains
    
    
    def _vertices(self, arg):
        if type(arg) is matplotlib.path.Path:
            vertices = arg.vertices
        elif type(arg) is XPath:
            vertices = arg.vertices
        else:
            vertices = arg
        return vertices

    
    def _path(self, arg):
        if type(arg) is XPath:
            path = arg
        else:
            path = XPath(arg)
        return path
            

def get_contour(X, Y, Z, threshold, index=-1):

    if not isiterable(index):
        index = [index]
        
    c = cntr.Cntr(X, Y, Z).trace(threshold)

    nseg = len(c) // 2
    segments = c[:nseg]
    lengths = np.asarray(sorted([(i, len(s))
                                 for i, s in enumerate(segments)],
                                key=lambda x: x[1]))
    idx = lengths[index][:,0]
    paths = [XPath(segments[i]) for i in idx]
    
    # remove nan's
    for i, path in enumerate(paths):
        ix = ~np.all(np.isnan(path.vertices), axis=1)
        paths[i] = XPath(path.vertices[ix,:])
        
    return paths


def get_paths(X, Y, Zs):

    paths = {}
    for j, (k, Z) in enumerate(Zs.iteritems()):
        paths[k] = {}

        c1 = get_contour(X, Y, Z, 0., index=[-1,-2,-3])
        c2 = get_contour(X, Y, Z, 3., index=[-1,-2])
        c3 = get_contour(X, Y, Z, 5., index=[-1,-2])
        paths[k]['contour_0m'] = c1
        paths[k]['contour_3m'] = c2
        paths[k]['contour_5m'] = c3
        
        # KNOWN SOURCES
        poly = zip(refx[[1,4,4,1]], refy[[1,1,4,4]])
        pth = c2[0].copy()
        pth.clip_to_polygon(poly)
        pth.extend(zip(refx[[1,4]], refy[[1,1]]))
        paths[k]['drybeach_center'] = pth

        # KNOWN SINKS
        poly = zip(refx[[2,2,3,4,4]], refy[[1,3,3,2,1]])
        pth = c2[0].copy()
        pth.clip_to_polygon(poly)
        paths[k]['lagoon'] = pth

        pth = c2[1].copy()
        paths[k]['lake'] = pth

        poly = zip(refx[[0,5,5,0]], refy[[0,0,2,2]])
        pth = XPath(poly)
        paths[k]['dunes'] = pth

        poly = zip(refx[[0,1,1,0]], refy[[0,0,2,2]])
        pth = XPath(poly)
        paths[k]['dunes_south'] = pth

        poly = zip(refx[[1,4,4,1]], refy[[0,0,2,2]])
        pth = XPath(poly)
        paths[k]['dunes_center'] = pth

        poly = zip(refx[[4,5,5,4]], refy[[0,0,2,2]])
        pth = XPath(poly)
        paths[k]['dunes_north'] = pth
        
        # UNKNOWNS
        poly = zip(refx[[0,2,2,3,4,4,5,5,0]], refy[[0,0,3,3,2,0,0,4,4]])
        pth1 = c1[0].copy()
        pth1.clip_to_polygon(poly)
        pth2 = c2[0].copy()
        pth2.clip_to_polygon(poly).reverse()
        pth = pth1.extend(pth2)
        
        dst = pth.get_distances()
        ixs = sorted([(i+1, d) for i, d in enumerate(dst) if ~np.isnan(d)], key=lambda x: x[1])
        ix1 = np.min((ixs[-1][0], ixs[-2][0]))
        ix2 = np.max((ixs[-1][0], ixs[-2][0]))

        pth1, pth2, pth3 = pth.split(ix1, ix2)

        if k.startswith('2011') or k.startswith('2012'):
            pth = pth3.extend(pth1).extend(pth2, merge=False)
        else:
            pth4 = c1[1].copy()
            pth4.clip_to_polygon(poly)
            pth = pth3.extend(pth1).extend(pth2).extend(pth4)
        paths[k]['intertidal'] = pth
        
        poly = zip(refx[[0,2,2,0]], refy[[0,0,-1,-1]])
        pth1 = pth.copy()
        pth1.clip_to_polygon(poly)
        paths[k]['intertidal_south'] = pth1

        poly = zip(refx[[2,-1,-1,2]], refy[[0,0,-1,-1]])
        pth2 = pth.copy()
        pth2.clip_to_polygon(poly)
        paths[k]['intertidal_north'] = pth2

        poly = zip(refx[[0,1,1,0]], refy[[1,1,2,2]])
        pth = c2[0].copy()
        pth.clip_to_polygon(poly)
        pth.extend(zip(refx[[0,1]], refy[[1,1]]))
        paths[k]['drybeach_south'] = pth

        poly = zip(refx[[4,5,5,4]], refy[[1,1,2,2]])
        pth = c2[0].copy()
        pth.clip_to_polygon(poly)
        pth.extend(zip(refx[[4,5]], refy[[1,1]]))
        paths[k]['drybeach_north'] = pth
        
    return paths


X, Y = rotate(grid['X'], grid['Y'], 48.)
paths = get_paths(X, Y, grid['Z'])

paths['_convexhull'] = {}
paths['_convexhull']['lake'] = compute_convexhull(paths, 'lake')
paths['_convexhull']['lagoon'] = compute_convexhull(paths, 'lagoon')
```

```python
def iterate_time(paths):
    for d, p in sorted(paths.iteritems()):
        try:
            yield datetime.strptime(d, '%Y%m%d'), d, p
        except:
            pass
        
        
def iterate_paths(paths):
    if type(paths) is matplotlib.collections.PathCollection:
        for path in paths.get_paths():
            yield path
    else:
        yield paths
```

```python
def interpolate_nans(X, Y, Z):
    ix = np.isnan(Z.flatten()).flatten()
    if np.sum(ix) > 0:
        Z[ix] = \
            scipy.interpolate.griddata((X.flatten()[~ix],
                                        Y.flatten()[~ix]), 
                                       Z.flatten()[~ix],
                                       (X.flatten()[ix],
                                        Y.flatten()[ix])).reshape(Z[ix].shape)
    return Z
    
    
def dataframe_normalize(df, t):
    df0 = dataframe_interp(df, t)
    return df.subtract(df0)


def dataframe_removeanomalies(df, anomalies=['20110904','20120621', '20121008', '20131001']): #,'20140702']):
    ix = [d for d in df.index if d.strftime('%Y%m%d') not in anomalies]
    return df.ix[ix]


def dataframe_interp(df, t):
    if type(df) is not pd.DataFrame:
        df = pd.DataFrame(df)
    return pd.Series([np.interp(t.toordinal(),
                                [x.toordinal() for x in df.index],
                                df[c])
                      for c in df.columns],
                     index=df.columns,
                     name=t)


def dataframe_getslope(df):
    o = [d.toordinal()/365.2425 for d in df.index]
    p = np.polyfit(o, df, 1)
    return p[0]


def compute_volumes(paths, X, Y, Z, grid_size=100.):
    
    keys = np.unique([k
                      for p in paths.itervalues()
                      for k, v in p.iteritems()
                      if type(p) is dict and type(v) is XPath])
    
    xy = np.concatenate((X.reshape((-1,1)),
                         Y.reshape((-1,1))), axis=1)
    
    ix = []
    area = {k:[] for k in keys}
    volume = {k:[] for k in keys}
    height = {k:[] for k in keys}
    change = {k:[] for k in keys}
    changep = {k:[] for k in keys}
    changen = {k:[] for k in keys}
    
    d_init = None
    d_prev = None

    # loop over time
    for i, (dt, d, path) in enumerate(iterate_time(paths)):
        
        if not Z.has_key(d):
            continue
        
        print d
        ix.append(dt)
        
        Zd = interpolate_nans(xy[:,0], xy[:,1], Z[d].flatten())
        
        # loop over areas
        for k in keys:

            # get indices in current area
            if '_convexhull' in paths.keys() and k in paths['_convexhull']:
                idx = paths['_convexhull'][k].contains_points(xy)
            else:
                idx = np.zeros((xy.shape[0],1), dtype=np.bool)
                for p in iterate_paths(path[k]):
                    idx = idx | p.contains_points(xy).reshape(idx.shape)
            
            idx = idx.flatten()
            Zk = Zd.reshape((-1,1))[idx]
            
            porfac = (1. - porosity[k]) / .6
            
            # compute area, volume and volume change
            area[k].append(np.nansum(idx) * grid_size)
            height[k].append(np.nanmean(Zk))
            volume[k].append(np.nansum(Zk) * grid_size * porfac)
            
            if d_prev is not None:
                dZ = Zk - Z[d_prev].reshape((-1,1))[idx]
                change[k].append(np.nansum(dZ) * grid_size * porfac)
            else:
                change[k].append(0.)
                
            if d_init is not None:
                dZ = Zk - Z[d_init].reshape((-1,1))[idx]
                dZp = np.maximum(0., dZ)
                dZn = np.minimum(0., dZ)
                changep[k].append(np.nansum(dZp) * grid_size * porfac)
                changen[k].append(np.nansum(dZn) * grid_size * porfac)
            else:
                changep[k].append(0.)
                changen[k].append(0.)
                d_init = d
                
        d_prev = d

    return (pd.DataFrame(area, index=ix),
            pd.DataFrame(height, index=ix),
            pd.DataFrame(volume, index=ix),
            pd.DataFrame(change, index=ix),
            pd.DataFrame(changep, index=ix),
            pd.DataFrame(changen, index=ix))
```
