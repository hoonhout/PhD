## Saltation

```python
from copy import deepcopy

def add_jump(axs, x, x0, h, w, n, level=0, cutoff=False, offspring=None, criticalfetch=1000):
    
    if x0 < criticalfetch:
        h = np.random.uniform(h, 3*h)
    else:
        h = np.random.uniform(.8*h, 1.2*h)
        
    w = np.random.uniform(w, 1.5*w)
    
    y = np.sqrt(1 - (2*x - 1)**2) * h * np.sqrt(1-x)
    
    if not cutoff or level < n:
        axs.plot(x0+w*x, y, '-k')
    else:
        axs.plot(x0+w*x[:25], y[:25], '--k')
        axs.scatter(x0+w*x[25], y[25], c='k')
    
    x0 += w*x[-1]
    
    if level < n:
        o = offspring
        if not o:
            o = np.random.randint(1,3)
        for i in range(o):
            add_jump(axs, x, x0, h, w, n, level=level+1, cutoff=cutoff, offspring=offspring, criticalfetch=criticalfetch)
    
    return x0, h, w

x = np.linspace(0,1,100)

# mess
fig, axs = plt.subplots()
for i in range(10):
    add_jump(axs, x, 0, .5, .8, 10, criticalfetch=10)
axs.set_xlim((0,30))
axs.set_ylim((0,100))
axs.set_xticks(())
axs.set_yticks(())
axs.set_xlabel('fetch')
axs.set_ylabel('height above bed')

# small fetches
fig, axs = plt.subplots()
for i in range(10):
    add_jump(axs, x, 0, .5, .8, 5)
axs.set_xlim((0,10))
axs.set_ylim((0,50))
axs.set_xticks(())
axs.set_yticks(())
axs.set_xlabel('fetch')
axs.set_ylabel('height above bed')

# single departing grain
fig, axs = plt.subplots()
add_jump(axs, x, 0, .5, .8, 5)
axs.set_xlim((0,10))
axs.set_ylim((0,50))
axs.set_xticks(())
axs.set_yticks(())
axs.set_xlabel('fetch')
axs.set_ylabel('height above bed')

# single step
fig, axs = plt.subplots()
add_jump(axs, x, 0, .5, .8, 1, cutoff=True, offspring=3)
axs.set_xlim((0,2))
axs.set_ylim((0,2))
axs.set_xticks(())
axs.set_yticks(())
axs.set_xlabel('fetch')
axs.set_ylabel('height above bed')
```

## Particles

```python
class Particles:
    
    x = []
    y = []
    vx = []
    vy = []
    inmotion = []
    
    s = []
    
    def __init__(self, n=10, v0=(5., 10.), alpha0=(30., 60.), inmotion=True):
        v0 = self.randomize(v0, n)
        alpha0 = self.randomize(alpha0, n)
        
        self.x = self.randomize((0., 10.), n)
        self.y = self.randomize(0., n)
        self.s = self.randomize((5., 20.), n)
        
        self.vx = v0 * np.cos(alpha0/180.*np.pi)
        self.vy = v0 * np.sin(alpha0/180.*np.pi)
        self.inmotion = self.randomize(inmotion, n)
        
        
    def update(self, dt, a=5., g=-9.81):
        
        ix = self.inmotion
        
        self.x[ix] += self.vx[ix] * dt
        self.y[ix] += self.vy[ix] * dt

        self.vx[ix] += a * dt
        self.vy[ix] += g * dt

        self.y[ix] = np.maximum(0., self.y[ix])

        ix = self.y == 0.
        
        self.vx[ix] = 0.
        self.vy[ix] = 0.
        self.inmotion[ix] = False
            
            
    def plot(self, ax=None):
        if not ax:
            ax = plt.gca()
        ax.scatter(self.x, self.y, s=self.s, c='k')
        
    
    @staticmethod
    def randomize(x, n):
        if type(x) in [list, tuple] and len(x) > 1:
            return np.asarray([x[0] + (x[1] - x[0]) * np.random.random() 
                               for i in range(n)])
        else:
            return np.asarray([x] * n)


        
       
p = Particles(n=100)
            
dt = .1
fig, axs = plt.subplots(figsize=(10,4))
for i in range(20):
    p.plot(ax=axs)
    p.update(dt)
    
axs.set_ylim((0, 5))
axs.set_aspect('equal')
axs.grid()
```

## Armoring

```python
root = '../Figures/'

def plot_bed(x, y, s, fname=None):

    fig, axs = plt.subplots(figsize=(10,4))
    axs.scatter(x, y, s=s, c='k')
    if y[0,0] < y[0,-1]:
        axs.plot([0,80], [-.5,7.5], '-k', linewidth=3)
    else:
        axs.plot([0,80], [7.5,-.5], '-k', linewidth=3)
    axs.set_xticks(())
    axs.set_yticks(())
    axs.set_axis_off()

    axs.set_xlim((0, 80))
    axs.set_ylim((-1,16))
    axs.set_aspect('equal')

    if fname:
        fig.savefig(os.path.join(root, '%s.pdf' % fname))
        
    return fig, axs


random_size = lambda m: m * f * np.random.random(x.shape)
sorted_size = lambda m: m * f * y0 / y0.max()

x, y0 = np.meshgrid(np.arange(0., 80., 1.),
                    np.arange(0., 7., 1.))
y = y0 + .1 * x
y2 = y0 - .1 * x + 8.

a = .5
h = 9.
px = a * (np.random.random(x.shape) - .5)
py = a * (np.random.random(x.shape) - .5)
x += px
y += py
y2 += py

f = 2.
m = 20.
s1 = random_size(m)
plot_bed(x, y, 0., fname='grains_none')
plot_bed(x, y, m/f, fname='grains_small')
plot_bed(x, y, m*f, fname='grains_large')
plot_bed(x, y, s1, fname='grains_random')

s2 = sorted_size(m)
s3 = .25*s1+.75*s2
plot_bed(x, y, s3, fname='grains_sorted')

xl = np.maximum(0., x[:1,:] - 30.) / 50.
s4 = .25*s1 + .75*(1.-xl)*s2
fig, axs = plot_bed(x, y, s4)
fig.savefig(os.path.join(root, 'grains_deposition.pdf'), transparent=True)

xy1 = [(80, -.5),(0, 7.5),(0, h),(80, h)]
fig, axs = plot_bed(x, y2, s4)
axs.add_patch(plt.Polygon(xy1, facecolor='blue', edgecolor='none', alpha=.25))
fig.savefig(os.path.join(root, 'grains_deposition2.pdf'))

xy1 = [(0, -.5),(80, 7.5),(80, h),(0, h)]
fig, axs = plot_bed(x, y, .25*s1+.75*s2)
axs.add_patch(plt.Polygon(xy1, facecolor='blue', edgecolor='none', alpha=.25))
fig.savefig(os.path.join(root, 'grains_waterlevel.pdf'))

L = 10.
a = 2.
wlx = x[:1,:]
xl = np.maximum(0., 30. - x) / 30.
wl = h + a * np.sin(wlx / L * 2 * np.pi) * xl[:1,:]
xy2 = np.concatenate((wlx, wl), axis=0).T
xy2 = np.concatenate((xy2, xy1[2::-1]))
s5 = (.25+.75*xl)*s1 + .75*(1.-xl)*s2
fig, axs = plot_bed(x, y, s5)
axs.add_patch(plt.Polygon(xy2, facecolor='blue', edgecolor='none', alpha=.25))
fig.savefig(os.path.join(root, 'grains_wave.pdf'))

fig, axs = plot_bed(x, y, s5)
axs.add_patch(plt.Polygon(xy2, facecolor='blue', edgecolor='none', alpha=.25))
for i in np.arange(0,9,1.5):
    axs.plot([0,80], [-.5+i,7.5+i], '-r')
for i in np.arange(0,81,10):
    axs.plot([i,i], [-.5+.1*i,7.+.1*i], '-r')
axs.add_patch(plt.Polygon([(60,7.),
                       (70,8.),
                       (70,9.5),
                       (60,8.5)], facecolor='r', edgecolor='none', alpha=.5))

xx = np.arange(-10,10,.1)
ax2 = fig.add_axes([.775,.2,.1,.2])
ax2.fill_between(xx, np.exp(-xx**2), color='r')
ax2.plot([-10, 10], [.4, .4], '-b', linewidth=3)
#ax2.annotate('moisture content', xy=(-2.,.2), color='b', fontweight='bold', fontsize=6, va='bottom')
ax2.set_xlim((-2,2))
ax2.set_ylim((0,1.))
ax2.set_xticks(())
ax2.set_yticks(())

plt.draw()

c = ax2.transAxes.transform((0.,1.))
c = axs.transData.inverted().transform(c)
axs.annotate('', xy=c, xytext=(65,8.25), arrowprops={'arrowstyle':'->', 'shrinkB':0})

fig.savefig(os.path.join(root, 'grains_grid.pdf'))
```

```python
gaussian = lambda x, mu, sigma: np.exp(-(x-mu)**2/sigma**2)

d50 = 10**np.linspace(np.log10(10), np.log10(1000), 100) * 1e-6
m = gaussian(d50, 300e-6, 200e-6) / 100.
uth = 100. * np.sqrt(9.81 * d50 * (2650. - 1.25)/2650.)
u = 8.
q = 1.5e-4 * (u - uth)**3

mx1 = []
mx2 = []

n = 200
fig, axs = plt.subplots(2, 1, figsize=(8,6), sharex=True)
axs = np.asarray([[axs[0], axs[0].twinx()],
                  [axs[1], axs[1].twinx()]])
for i in range(n):
    p = m/np.sum(m)
    if np.mod(i, 30) == 0:
        c = 1. - i/float(n)
        lw = 5.*i/n
        #h1 = axs[0,1].plot(d50*1e6, p, color=(c,c,c), linewidth=lw, label='sediment in bed')
        axs[0,0].fill_between(d50*1e6, np.zeros(d50.shape), p, facecolor=(c,c,1.), edgecolor='none')
        axs[1,0].fill_between(d50*1e6, np.zeros(d50.shape), q * p, facecolor=(1.,c,c), edgecolor='none')
        mx1.append((d50[np.argmax(p)], np.max(p)))
        mx2.append((d50[np.argmax(q*p)], np.max(q*p)))
    m = np.maximum(0., m - q * p)
    
o = .02
axs[0,0].annotate('time', xy=(mx1[1][0]*1e6, 1.5*o+mx1[1][1]), color='b', rotation=30, ha='left', va='bottom')
axs[0,0].annotate('', xy=(mx1[-1][0]*1e6, o+mx1[-1][1]), xytext=(mx1[1][0]*1e6, o+mx1[1][1]), 
                  color='b', arrowprops={'arrowstyle':'->', 'shrinkB':0, 'linewidth':3, 'color':'b'})

o = 2e-5
axs[1,0].annotate('time', xy=(mx2[1][0]*1e6, .25*o+mx2[1][1]), color='r', rotation=-45, ha='left', va='bottom')
axs[1,0].annotate('', xy=(mx2[-1][0]*1e6, o+mx2[-1][1]), xytext=(mx2[1][0]*1e6, o+mx2[1][1]), 
                  color='r', arrowprops={'arrowstyle':'->', 'shrinkB':0, 'linewidth':3, 'color':'r'})

for ax in axs[:,1]:
    h1 = ax.plot(d50*1e6, q, '-g', linewidth=3., label='equilibrium sediment transport')
    ax.set_ylim((0,.05))
    
for ax in axs[:,0]:
    ax.grid()

for ax in axs.flatten():
    ax.set_yticks(())

axs[1,0].set_xlabel('grain size [$\mathrm{\mu m}$]')
axs[0,0].set_ylim((0,.35))
axs[1,0].set_ylim((0,2e-4))

h2 = matplotlib.patches.Patch(color='b', label='normalized sediment availibility')
h3 = matplotlib.patches.Patch(color='r', label='sediment transport')
axs[0,0].legend(handles=(h1[0],h2,h3), loc='upper right')

fig.savefig('../Figures/transport.pdf')
```
