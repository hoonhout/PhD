```python
def interpolate_nans(X, Y, Z):
    ix = np.isnan(Z.flatten()).flatten()
    if np.sum(ix) > 0:
        Z[ix] = \
            scipy.interpolate.griddata((X.flatten()[~ix],
                                        Y.flatten()[~ix]), 
                                       Z.flatten()[~ix],
                                       (X.flatten()[ix],
                                        Y.flatten()[ix])).reshape(Z[ix].shape)
    return Z
    
    
def dataframe_normalize(df, t):
    df0 = dataframe_interp(df, t)
    return df.subtract(df0)


def dataframe_removeanomalies(df, anomalies=['20110904','20120621', '20121008', '20131001']): #,'20140702']):
    ix = [d for d in df.index if d.strftime('%Y%m%d') not in anomalies]
    return df.ix[ix]


def dataframe_interp(df, t):
    if type(df) is not pd.DataFrame:
        df = pd.DataFrame(df)
    return pd.Series([np.interp(t.toordinal(),
                                [x.toordinal() for x in df.index],
                                df[c])
                      for c in df.columns],
                     index=df.columns,
                     name=t)
```

```python
import netCDF4
from datetime import datetime, timedelta

def get_volume_change(ncfile, n=0):#, paths):
    
    #keys = sorted(paths.keys())
    #paths_model = paths.copy()
    with netCDF4.Dataset(ncfile, 'r') as ds:

        t = ds.variables['time'][n:]
        Xm = ds.variables['x'][:,:]
        Ym = ds.variables['y'][:,:]
        zb = ds.variables['zb'][n:,:,:]

        dZ = -ds.variables['pickup.sum'][n:,:,:,:].sum(axis=-1).cumsum(axis=0) / 2650. / .6
        
        constant_bathy = np.diff(zb, axis=0).sum() == 0.

        Zm = {}
        dZm = {}
        for i in range(0, len(t), 1):
            dt = t0 + timedelta(seconds=t[i])
            #ix = np.argmin(np.abs(np.asarray([datetime.strptime(ts, '%Y%m%d') 
            #                                  for ts in keys 
            #                                  if not ts.startswith('_')]) - dt))
            k = dt.strftime('%Y%m%d')
            dZm[k] = dZ[i,:,:]
            dZm[k][:,0] = -dZ[i,:,1:].sum(axis=-1)
            Zm[k] = zb[i,:,:] 
            
            #if constant_bathy:
                #paths_model[k] = paths['20110803']
                #paths_model[k] = paths['20150922']
            #else:
                #paths_model[k] = paths[keys[ix]]
            #paths_model[k] = paths['20150922']
            
    return Xm, Ym, Zm, dZm #, paths_model
```

```python
def aggregate_volumes(Xm, Ym, Zm, dZ):
    
    V = dict(lagoon = [],
             lake = [],
             dunes = [],
             intertidal = [],
             drybeach_center = [],
             drybeach_center_height = [],
             drybeach_north = [],
             drybeach_south = [],
             offshore = [],
             deficit = [])
    idx = []

    for k in sorted(Zm.keys()):
        idx.append(datetime.strptime(k, '%Y%m%d'))
        
        # determine dune area
        j, i = np.where(Zm[k]<5.)
        ix = pd.DataFrame(i).groupby(j).min().values
        ix_dunes = Xm <= ix.repeat(Xm.shape[1], axis=1)
        
        ix1 = (Ym >= 200.) & (Ym < 1200.) & (Xm >= -300.) & (Xm < 200.) & (dZ[k] > 0.)
        V['lagoon'].append(dZ[k][ix1].sum() * 2500.)
        
        ix2 = (Ym >= -500.) & (Ym < 200.) & (Xm >= -300.) & (Xm < 0.) & (dZ[k] > 0.)
        V['lake'].append(dZ[k][ix2].sum() * 2500.)

        ix3 = (Xm < -300.) & (dZ[k] > 0.)
        V['dunes'].append(dZ[k][ix3].sum() * 2500.)

        ix4 = (Xm >= -450.) & (Zm[k] < 3.) & (dZ[k] < 0.)
        V['intertidal'].append(dZ[k][ix4].sum() * 2500.)
        
        ix5 = (Xm >= -450.) & (Ym >= -1000) & (Ym < 1200.) & (Zm[k] >= 3.) & (dZ[k] < 0.)
        V['drybeach_center'].append(dZ[k][ix5].sum() * 2500.)

        ix6 = (Xm >= -450.) & (Ym >= -1000) & (Ym < 1200.) & (Zm['20150827'] >= 3.) & (dZ[k] < 0.) & ~ix_dunes
        V['drybeach_center_height'].append((Zm[sorted(Zm.keys())[0]] + dZ[k])[ix6].mean())

        ix7 = (Xm >= -450.) & (Ym >= 1200.) & (Zm[k] >= 3.) & (dZ[k] < 0.)
        V['drybeach_north'].append(dZ[k][ix7].sum() * 2500.)

        ix8 = (Xm >= -450.) & (Ym < -1000.) & (Zm[k] >= 3.) & (dZ[k] < 0.)
        V['drybeach_south'].append(dZ[k][ix8].sum() * 2500.)

        ix9 = (Xm >= -450.) & ( (Xm >= 200.) | (Ym < -500.) | (Ym >= 1200.) ) & (dZ[k] > 0.)
        V['offshore'].append(dZ[k][ix9].sum() * 2500.)

        V['deficit'].append((dZ[k]).sum() * 2500.)

    volumes = pd.DataFrame(V, index=idx)

    volumes.ix[-1] / 1e4
    
    return volumes
```

```python
def plot_volumes_ts(volumes, t0, t1):
    
    a1 = dataframe_removeanomalies(dataframe_normalize(volumes, t0).divide(10000.))
    
    sinks = a1.filter(['lagoon', 'lake', 'dunes']).sum(axis=1)
    sources = a1.filter(['drybeach_center', 'drybeach_south', 'drybeach_north']).sum(axis=1)
    deficit = -a1.filter(['intertidal', 'offshore']).sum(axis=1)
    
    fig, axs = subplots(figsize=(10,4))
    axs.plot(sinks.index, sinks, '-r', label='modeled deposition')
    axs.plot(sources.index, sources, '-g', label='modeled erosion')
    axs.plot(deficit.index, deficit, '-b', label='modeled deficit')
    
    axs.set_xlim((t0, t1))
    axs.set_ylim((-20, 60))
    axs.grid()
    
    axs.set_xlabel('time [years]')
    axs.set_ylabel('cumulative volume change [$\mathrm{10.000 m^3}$]')
    axs.xaxis.set_major_locator(matplotlib.dates.YearLocator())
    
    return (fig, axs), (sources, sinks, deficit)
```

```python
def align_ts(data1, data2):
    o1 = [ix.toordinal() for ix in data1.index]
    o2 = [ix.toordinal() for ix in data2.index]
    return np.interp(o1, o2, data2)


def compute_rmse(data1, data2):
    rmse = np.sqrt(np.mean((align_ts(data1, data2) - data1)**2))

    return rmse

def compute_r2(data1, data2, p=2):
    SStot = np.sum((data1 - data1.mean())**2)
    SSres = np.sum((align_ts(data1, data2) - data1)**2)
    r2 = 1. - SSres/SStot
    
    #n = len(data1)
    #r2 = 1.-(1.-r2)*(n-1)/(n-p) # adjusted r2

    return r2
```

# Compute error statistics

```python
import re

ERR = {}

def compute_weightedavg(err, key, weights={'combined': 1., 'sinks': 1.}):
#def compute_weightedavg(err, key, weights={'sources': 1., 'deficit':1., 'sinks': 2.}):
    return np.sum([err[k][key] * w/np.sum(weights.values()) for k, w in weights.iteritems()])

for ncfile, v in VOL.iteritems():
    
    sigma, Tdry = [float(x) for x in re.search('sandmotor_sigma([\d\.]+)_Tdry([\d\.]+).*\.nc$', ncfile).groups()]
    
    if Tdry > 10 or sigma > 20:
        continue
        
    measured = v['measured']
    modeled = v['modeled']
    
    i0 = 0
    ERR[ncfile] = {'sources':{'RMSE':compute_rmse(measured['sources'].ix[i0:], modeled['sources'].ix[i0:]),
                              'R2':compute_r2(measured['sources'].ix[i0:], modeled['sources'].ix[i0:])},
                   'sinks':{'RMSE':compute_rmse(measured['sinks'].ix[i0:], modeled['sinks'].ix[i0:]),
                            'R2':compute_r2(measured['sinks'].ix[i0:], modeled['sinks'].ix[i0:])},
                   'deficit':{'RMSE':compute_rmse(measured['deficit'].ix[i0:], modeled['deficit'].ix[i0:]),
                              'R2':compute_r2(measured['deficit'].ix[i0:], modeled['deficit'].ix[i0:])},
                   'combined':{'RMSE':compute_rmse(measured['sources'].ix[i0:]-measured['deficit'].ix[i0:],
                                                   modeled['sources'].ix[i0:]-modeled['deficit'].ix[i0:]),
                              'R2':compute_r2(measured['sources'].ix[i0:]-measured['deficit'].ix[i0:],
                                              modeled['sources'].ix[i0:]-modeled['deficit'].ix[i0:])}}
    
    ERR[ncfile]['avg'] = {'RMSE':compute_weightedavg(ERR[ncfile], 'RMSE'),
                          'R2':compute_weightedavg(ERR[ncfile], 'R2')}
    
    ERR[ncfile]['ratio'] = modeled['deficit'].ix[-1] / (modeled['deficit'].ix[-1] - modeled['sources'].ix[-1])
```
