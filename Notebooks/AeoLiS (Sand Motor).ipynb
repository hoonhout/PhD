```python
def rotate(x, y, phi, origin=(72502.0, 452074.0)):

    x = np.asarray(x)
    y = np.asarray(y)
    phi = phi / 180. * np.pi

    R = np.asarray([[np.cos(phi), -np.sin(phi)],
                    [np.sin(phi),  np.cos(phi)]])

    XY = np.concatenate((x.reshape(-1,1) - origin[0],
                         y.reshape(-1,1) - origin[1]), axis=1)

    XY = np.dot(XY, R)
    xr = XY[:,0].reshape(x.shape)
    yr = XY[:,1].reshape(y.shape)
    
    return xr, yr


# select and rotate bathy
alpha = 48.
X, Y = rotate(grid['X'], grid['Y'], alpha)
Z = grid['Z'][key]
```

```python
# interpolate bathy to computational grid and write to file

d = 50
x = np.arange(-500,  1000, d)
y = np.arange(-1500, 2500, d)

Xr, Yr = np.meshgrid(x, y)
np.savetxt(os.path.join(root, '..', 'x.txt'), Xr)
np.savetxt(os.path.join(root, '..', 'y.txt'), Yr)

for i, key in enumerate(sorted(grid['Z'].keys())):
    Z = grid['Z'][key]
    
    ix = ~np.isnan(Z.flatten())
    Zr = scipy.interpolate.griddata((Y.flatten()[ix],
                                     X.flatten()[ix]),
                                    Z.flatten()[ix],
                                    (Xr.flatten(),
                                     Yr.flatten()), method='nearest').reshape(Xr.shape)

    np.savetxt(os.path.join(root, '..', 'bathy', 'z_%s.txt' % key), Zr)

print Zr.shape
```

```python
# compute tidal masks
for fname in glob.glob(os.path.join(root, '..', 'bathy', 'z_*.txt')):
    
    Z = np.loadtxt(fname)
    
    # get contour
    yi, xi = np.where(Z > 3.)
    ix = pd.DataFrame({'x':xi,'y':yi}).groupby('y').max().values.flatten()

    # convert to mask
    mask1 = 10+Xr//50 < ix.reshape((-1,1)).repeat(30,axis=1)
    mask1 = 1. - mask1
    
    # get tangent
    ix[45:] = np.interp(np.arange(45,80),[45,79],ix[[45,79]])

    # convert to mask
    mask2 = 10+Xr//50 < ix.reshape((-1,1)).repeat(30,axis=1)
    mask2 = 1. - mask2
    
    # interpolate transition
    mask = mask1.copy()
    ix = (mask1 == mask2)
    F = scipy.interpolate.LinearNDInterpolator(np.concatenate((Xr[ix].reshape((-1,1)),
                                                               Yr[ix].reshape((-1,1))), axis=1),
                                               mask[ix].reshape((-1,1)))
    mask[~ix] = F(Xr[~ix], Yr[~ix]).flatten()
    
    # add complex component
    mask = np.asarray(mask, dtype=np.complex)
    mask += (1. - mask) * 1j
    
    np.savetxt(fname.replace('z_', 'mask_'), mask)
```

```python
def create_plot(par, data):
    
    # remove inf's
    ix = np.isinf(data)
    data[ix] = np.nan
    
    # remove extra dimensions
    if data.ndim == 4:
        data = data[:,:,0,:] # pick top layer
    if data.ndim == 3:
        data = data.sum(axis=-1) # sum over fractions
    
    # determine data range
    mn = np.nanmin(data[1:-1,1:-1])
    mx = np.nanmax(data[1:-1,1:-1])
    mx = np.maximum(np.abs(mn), np.abs(mx))

    if mn > -1e-10:
        mn = 0.
        cmap = 'Reds'
    else:
        mn = -mx
        cmap = 'bwr'

    # create plot
    fig, axs = subplots(figsize=(10,4))
    p = axs.pcolormesh(y, x, data, cmap=cmap, vmin=mn, vmax=mx)
    axs.set_aspect('equal')
    axs.set_title(par)
    fig.colorbar(p)

    # save plot
    fig.savefig(os.path.join(root, 'plots', '%s.png' % par))

    
n = 10 # number of time steps
with netCDF4.Dataset(os.path.join(root, 'aeolis.nc'), 'r') as ds:
    t = ds.variables['time'][:] / 3600.
    x = ds.variables['x'][:,:]
    y = ds.variables['y'][:,:]
    f = ds.variables['fractions'][:]
    
    nx = x.shape[1]
    ny = y.shape[0]
    
    for t in range(5): #np.linspace(1, len(t)-1, n):
        
        t = int(np.round(t))
    
        # bathy difference
        data = ds.variables['zb'][...]
        create_plot('dz_t=%d' % t, data[t,:,:] - data[0,:,:])

        # mean grain size
        data = ds.variables['mass'][...]
        gs = ds.groups['settings'].getncattr('grain_size') * 1e6
        create_plot('d50_t=%d' % t,
                    np.asarray([np.average(gs, weights=ds.variables['mass'][t,j,i,0,:])
                                for j in range(ny) for i in range(nx)]).reshape((ny, nx)))
        
        # other variables
        for par in ['Ct', 'Cu', 'pickup', 'zb', 'zs']:
            create_plot('%s_t=%d' % (par, t), ds.variables[par][t,...])
```

```python
import matplotlib.animation as animation
from datetime import datetime, timedelta

def rotate(x, y, phi, origin=(0, 0)):

    x = np.asarray(x)
    y = np.asarray(y)
    phi = phi / 180. * np.pi

    R = np.asarray([[np.cos(phi), -np.sin(phi)],
                    [np.sin(phi),  np.cos(phi)]])

    XY = np.concatenate((x.reshape(-1,1) - origin[0],
                         y.reshape(-1,1) - origin[1]), axis=1)

    XY = np.dot(XY, R)
    xr = XY[:,0].reshape(x.shape)
    yr = XY[:,1].reshape(y.shape)

    return xr, yr
    
    
class AeoLiSAnimation(matplotlib.animation.FuncAnimation):
    
    
    t0 = datetime(2011,9,1)
    
    
    def __init__(self, figsize=(15, 8), frames=3, blit=False):
        
        fig = plt.figure(figsize=figsize, facecolor='white')
        
        gs = matplotlib.gridspec.GridSpec(6, 2, hspace=1)
        
        axs = []
        axs.append(plt.subplot(gs[:2,0]))
        axs.append(plt.subplot(gs[:2,1]))
        axs.append(plt.subplot(gs[2:4,0]))
        axs.append(plt.subplot(gs[2:4,1]))
        axs.append(plt.subplot(gs[4,:]))
        axs.append(plt.subplot(gs[5,:]))

        self.fig = fig
        self.axs = np.asarray(axs)
        
        self.axp = fig.add_axes([0,0,1,.025])
        self.axp.set_axis_off()
        self.axp.set_xlim((0,1))

        super(AeoLiSAnimation, self).__init__(self.fig, self.update,
                                              frames=frames, init_func=self.initialize, blit=blit)
        
        
    def initialize(self):
        
        x = data1['x']
        y = data1['y']

        self.meshes = [-1] * 4
        self.meshes[0] = self.axs[0].pcolormesh(y, x, np.zeros(x.shape), cmap=cmap, vmin=-12.5, vmax=7.5)
        self.meshes[1] = self.axs[2].pcolormesh(y, x, np.zeros(x.shape), cmap='bwr', vmin=-2, vmax=2)
        self.meshes[2] = self.axs[1].pcolormesh(y, x, np.zeros(x.shape), cmap=cmap, vmin=-12.5, vmax=7.5)
        self.meshes[3] = self.axs[3].pcolormesh(y, x, np.zeros(x.shape), cmap='bwr', vmin=-2, vmax=2)
        
        self.axs[4].plot([self.t0 + timedelta(seconds=t) for t in tide[:,0]], tide[:,1], '-k')
        self.axs[5].plot([self.t0 + timedelta(seconds=t) for t in waves[:,0]], waves[:,1], '-k')
        
        self.progress1 = self.axs[4].plot([self.t0, self.t0], [-10, 10], '-r')[0]
        self.progress2 = self.axs[5].plot([self.t0, self.t0], [-10, 10], '-r')[0]

        for ax in self.axs[:2]:
            ax.set_xticklabels(())

        for ax in self.axs[[1,3]]:
            ax.set_yticklabels(())

        for ax in self.axs[2:4]:
            ax.set_xlabel('alongshore distance [m]')

        for ax in self.axs[[0,2]].flatten():
            ax.set_ylabel('cross-shore distance [m]')

        for ax in self.axs[:4].flatten():
            ax.set_xlim((-1450, 2450))
            ax.set_ylim((-500, 950))
            ax.set_aspect('equal')
            ax.grid()
            
        for ax in self.axs[4:-1]:
            ax.set_xticklabels(())
            
        self.axs[0].set_title('single sediment fraction')
        self.axs[1].set_title('multiple sediment fractions')
        
        self.axs[4].set_ylabel('water level\n[m+MSL]')
        self.axs[4].set_ylim((-1.5, 2.5))
        self.axs[4].set_yticks((-1.5, 0, 2.5))
        
        self.axs[5].set_ylabel('wave height\n[m]')
        self.axs[5].set_ylim((0, 6))
        self.axs[5].set_yticks((0, 3, 6))
        
        self.axs[-1].set_xlabel('time [years]')
        
        self.fig.colorbar(self.meshes[0], shrink=1., ax=list(self.axs[:2]))
        self.fig.colorbar(self.meshes[1], shrink=1., ax=list(self.axs[2:4]))
        cb = self.fig.colorbar(self.meshes[1], shrink=1., ax=list(self.axs[4:]))
        cb.ax.set_axis_off()
        cb.outline.set_edgecolor('w')
        cb.ax.add_patch(Rectangle((0,0), 1, 1, facecolor='w', edgecolor='w'))
        
        #self.progressbar = self.axp.add_patch(Rectangle((0,0), 0, 1, color='k'))
        #self.progresslabel = self.axp.annotate('2011', xy=(0, 1), va='bottom', ha='left',
        #                                       fontsize=16, fontweight='bold', color='k')
        
        verts = self.blockarrow((2000, 600), scale=100, alpha=48.)
        self.axs[3].add_patch(Polygon(verts[:-1,:], facecolor='k', edgecolor='k'))
        self.axs[3].annotate('N', xy=(2200, 750), fontsize=10, fontweight='bold', ha='center', va='center')
        
        verts = self.blockarrow((-1000, 600), scale=100, alpha=0.)
        self.windarrow = self.axs[2].add_patch(Polygon(verts[:-1,:], facecolor='k', edgecolor='k'))
        
        self.update(0)


    def update(self, i):
    
        self.set_mesh(self.meshes[0], data1['zb'][i,...])
        self.set_mesh(self.meshes[1], data1['zb'][i,...]-data1['zb'][0,...])
        self.set_mesh(self.meshes[2], data2['zb'][i,...])
        self.set_mesh(self.meshes[3], data2['zb'][i,...]-data2['zb'][0,...])
        
        #p = float(i)/(len(data1['t'])-1)
        #self.progressbar.set_width(p)
        #self.progresslabel.set_text((datetime(2011,9,1) + timedelta(seconds=i*3600.*24.*365.2425/12.)).strftime('%Y'))
        #self.progresslabel.set_position((p, 1))
        t = self.t0 + timedelta(seconds=i*3600.*24.*365.2425/12.)
        self.progress1.set_data([t,t], [-10,10])
        self.progress2.set_data([t,t], [-10,10])
        
        verts = self.blockarrow((-1000, 600), scale=20*data1['uw'][i,0,0], alpha=data1['udir'][i,0,0])
        self.windarrow.set_xy(verts[:-1,:])
        
        return p
        
                
    @staticmethod
    def set_mesh(mesh, data):
        arr = mesh.get_array()
        arr[:] = data[:-1,:-1].reshape(arr.shape)
        mesh.set_array(arr)
        
        
    @staticmethod
    def blockarrow(xy, scale, alpha):    
        verts = np.asarray([[ 0., -.5],
                            [-.7, -1.],
                            [ 0.,  1.],
                            [ .7, -1.],
                            [-.15, 1.5]]).T

        x, y = rotate(verts[0,:], 
                      verts[1,:], 
                      alpha)

        verts = np.concatenate((xy[0]+scale*x, xy[1]+scale*y)).reshape(verts.shape).T
        return verts

    
im_ani = AeoLiSAnimation(frames=len(data1['t']), blit=False)
im_ani.save('aeolis_sandmotor.mp4', fps=10, dpi=100)
```
