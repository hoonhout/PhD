```python
fig, axs = subplots(figsize=(7,7))

r = 'h'

map = Basemap(projection='merc',
              llcrnrlat=51.5,
              urcrnrlat=52.5,
              llcrnrlon=3.5,
              urcrnrlon=5,
              resolution=r)
map.drawcoastlines(ax=axs)
map.fillcontinents(ax=axs) #, color=(.9,.9,0.), lake_color='white')
map.drawmapscale(4.7, 51.55, 4.25, 52, 50, ax=axs)

#locs = OrderedDict([('Kijkduin',   (52.069857, 4.221302)),
#                    ('Sand Motor', (52.053359, 4.185319)),
#                    ('Vlugtenburg',(52.002304, 4.128699))])
locs = OrderedDict([('Sand Motor', (52.053359, 4.185319))])

for i, (s, (lat, lon)) in enumerate(locs.iteritems()):
    ha = 'right' if mod(i,2) else 'left'
    offset = -1 if mod(i,2) else 1
    
    x, y = map(lon, lat)
    axs.scatter(x, y, s=100, c='r', label=s, zorder=100)
    axs.annotate(s, (x, y),
                 xytext=(10 * offset, 0), textcoords='offset points',
                 va='center', ha=ha)
    
map2 = Basemap(projection='merc',
              llcrnrlat=50,
              urcrnrlat=55,
              llcrnrlon=-1,
              urcrnrlon=7,
              resolution=r)

ax2 = axes([.175, .6, .275, .275])
map2.drawcoastlines(ax=ax2)
map2.fillcontinents(ax=ax2)
map2.drawmapscale(6.4, 51.3, 4.25, 52, 100, ax=ax2)

lat, lon = locs['Sand Motor']
x, y = map2(lon, lat)
ax2.scatter(x, y, s=100, c='r', zorder=100)

fig.savefig('../Figures/map.pdf')
```

# Maps

```python
def mark_corners(m, xy, w, h, p=.1, ax=None, **kwargs):
    x0 = xy[0,0] * np.ones((2,))
    x1 = xy[1,0] * np.ones((2,))
    y0 = xy[0,1] * np.ones((2,))
    y1 = xy[1,1] * np.ones((2,))

    m.plot(x0 + [0, p*w], y0, '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x1 - [0, p*w], y0, '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x0 + [0, p*w], y1, '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x1 - [0, p*w], y1, '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x0, y0 + [0, p*h], '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x0, y1 - [0, p*h], '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x1, y0 + [0, p*h], '-k', latlon=True, ax=ax, **kwargs)
    m.plot(x1, y1 - [0, p*h], '-k', latlon=True, ax=ax, **kwargs)
```

```python
xy1 = np.asarray([(-10., 48.),
                  (12., 60.)])
xy2 = np.asarray([(2., 51.),
                  (6., 53.5)])
xy3 = np.asarray([(4.1, 52),
                  (4.3, 52.1)])

clr = (.5,.5,.5)
fsize = (5, 5)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

fig, axs = subplots(figsize=fsize)

m = Basemap(projection='ortho', resolution=None, lat_0=52.053359, lon_0=4.185319, ax=axs)
m.shadedrelief(ax=axs)

parallels = np.arange(0., 180., 20.)
meridians = np.arange(0., 360., 30.)

m.drawparallels(parallels, color=clr, textcolor=clr, ax=axs)
m.drawmeridians(meridians, color=clr, textcolor=clr, ax=axs)

mark_corners(m, xy1, np.diff(xy1[:,0]), np.diff(xy1[:,1]), p=.2, ax=axs)

for p in axs.patches:
    p.set_edgecolor(clr)
    
fig.savefig('../Figures/map_earth.pdf')
```
