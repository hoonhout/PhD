```python
import os
import pytz
import netCDF4
import pandas as pd
from datetime import datetime

def read_data(ncfile, datefrom, dateto, freq='10Min', root='/Users/hoonhout/Checkouts/Zandmotor/aeolian/megapex/results/'):
    
    ncfile = os.path.join(root, ncfile)

    if not os.path.exists(ncfile):
        raise IOError('File not found [%s]' % ncfile)
    dfs = []
    
    try:
        nc = netCDF4.Dataset(ncfile, 'r')
        t = netCDF4.num2date(nc.variables['time'][:],
                             nc.variables['time'].units)

        print zip(nc.variables['x'][:], nc.variables['y'][:])
        
        ix1 = np.maximum(0, np.argmin(np.abs(t-datefrom))-1)
        ix2 = np.argmin(np.abs(t-dateto))+1
        
        n = len(nc.dimensions['timeSeries'])
        for i in range(n):
            dfs.append({})
            for key, var in nc.variables.iteritems():
                if 'time' in var.dimensions:
                    if 'timeSeries' in var.dimensions:
                        dfs[i][key] = var[ix1:ix2,i]
                    elif key not in nc.dimensions:
                        dfs[i][key] = var[ix1:ix2]
            dfs[i] = pd.DataFrame(dfs[i], index=t[ix1:ix2]).ix[datefrom:dateto].resample(freq, base=0)

    finally:
        nc.close()
        
    return dfs
```

```python
from scipy.optimize import curve_fit

def plot_data(ncfile, t1, t2):
    dfs = read_data(ncfile, t1, t2)
    dt1 = t1.strftime('%Y%m%d%H%M%S')
    dt2 = t2.strftime('%Y%m%d%H%M%S')
    fname = ncfile.replace('deployment_', '') \
                  .replace('_merged.nc', '_%s_%s' % (dt1, dt2))
    for i, df in enumerate(dfs):
        for c in df.columns:
            if 'direction' in c:
                df[c] = df[c].apply(lambda x: mod(x, 360))
        
        plot_series(df, fname='timeseries_%s_%d' % (fname, i))
        plot_fit(df, fname='distribution_%s_%d' % (fname, i))
        
    plot_correlations(dfs, fname='correlations_%s' % fname)
    

def plot_series(df, fname=None):
    n = len(df.columns)
    nx = int(np.ceil(np.sqrt(n)))
    ny = int(np.ceil(n/float(nx)))

    fig, axs = subplots(ny, nx, figsize=(20,20))
    axs = axs.flatten()
    for j, col in enumerate(df.columns):
        df.filter([col]).plot(legend=False, ax=axs[j], style='-ok')
        axs[j].set_title(col)
#            plt.setp(axs[j].xaxis.get_ticklabels(), rotation=40)

    fig.tight_layout(rect=[0,0,1,.96])

    if fname is not None:
        fig.text(.5, .98, fname, ha='center', va='bottom')
        fig.savefig('%s.png' % fname)
        
    plt.close(fig)


def plot_fit(df, freq='10Min', fname=None):

    def func_lin(x, a, b):
        return a*x + b

    def func_log(x, a, b, c):
        return a + log(b*x+c)

    def func_exp(x, a, b):
        return a*exp(b*x)

    df = df.filter(like='particle_count').resample(freq, base=0)
    df = df.div(df.sum(axis=1), axis=0)

    fig, axs = subplots(2, 3, figsize=(20,10))

    h = np.asarray([0.03, 0.10, 0.25, 0.40, 0.55, 0.70][:len(df.columns)])

    fits = []
    fcns = [func_lin, func_log, func_exp]
    for i, fcn in enumerate(fcns):
        for j, i0 in enumerate([0, 1]):
            if len(h) - i0 < 3:
                continue
            p, c = curve_fit(fcn, h[i0:], df.mean()[i0:])
            fits.append({'fcn':fcn,
                         'par':p,
                         'err':np.sqrt(np.mean((fcn(h[i0:], *p) - df.mean()[i0:])**2))})
            
            axs[j,i].plot(df.T, h, '-', color=(.9,.9,.9))
            axs[j,i].plot(df.mean(), h, '-or', linewidth=3, markeredgecolor='r')
            axs[j,i].plot(df.mean()-df.std(), h, ':r')
            axs[j,i].plot(df.mean()+df.std(), h, ':r')

            axs[j,i].plot(fits[-1]['fcn'](h, *fits[-1]['par']), h, '-g', linewidth=3)
            axs[j,i].set_title('%s (RMSE: %f)' % (fcn.__name__, fits[-1]['err']))
            axs[j,i].set_xlim((0, 1))
            axs[j,i].set_xlabel('count fraction [-]')
            axs[j,i].set_ylabel('height above bed [m]')
            
    fig.tight_layout(rect=[0,0,1,.96])
    
    if fname is not None:
        fig.text(.5, .98, fname, ha='center', va='bottom')
        fig.savefig('%s.png' % fname)
            
    plt.close(fig)
```

```python
def group_correlations(x):
    if x.startswith('particle_count'):
        return 'particle_count'
    elif 'wind_speed' in x or 'wind_velocity' in x:
        return 'wind_speed'
    elif 'direction' in x:
        return 'wind_direction'
    elif x.startswith('zandmotor_'):
        return x.replace('zandmotor_', '')
    else:
        return x

    
def plot_correlations(dfs, fname=None):
    
    fig, axs = subplots(len(dfs), 3, figsize=(20,20))

    for i, df in enumerate(dfs):
        
        c = df.corr()

        for j, flt in enumerate([(0,3), (3,10), (0,10)]):
            s = c.filter(['particle_count_%d' % (ind+1) for ind in range(*flt)]).mean(axis=1)
            g = s.groupby(by=group_correlations)

            df = pd.DataFrame({'avg':g.mean(),
                               'std':g.std(),
                               'abs':abs(g.mean())}, index=g.mean().index).sort('abs', ascending=False)

            df.filter(['avg']).plot(ax=axs[i,j], kind='bar', legend=False, color='y', yerr=df['std'])

            axs[i,j].set_ylim((-.5, 1))
            axs[i,j].set_ylabel('correlation coefficient')
            axs[i,j].set_title('location #%d, sensors %d-%d' % (i+1, flt[0], flt[1]))

            plt.setp(axs[i,j].get_xticklabels(), rotation=30, ha='right')

    fig.tight_layout(rect=[0,0,1,.96])
    
    if fname is not None:
        fig.text(.5, .98, fname, ha='center', va='bottom')
        fig.savefig('%s.png' % fname)
        
    plt.close(fig)
```
