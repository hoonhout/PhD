# Read meteo

```python
meteo, legend = knmi.read_uurgeg('../Data/knmi/uurgeg_330_2011-2020.txt')
ix = (meteo['DD'] > 360.) | (meteo['FH'] >= 900)
meteo = meteo.ix[~ix]
u10 = meteo['FH'] / 10.
```

# Compute threshold shear velocity

```python
d = 335e-6
alpha = .174/np.log10(10./.01)
uth = .085 * np.sqrt((2650.-1.25)/1.25 * 9.81 * d) / alpha
print alpha, uth, uth*alpha
```

# Alongshore transport

```python
df_dir = np.sin(np.radians(312.-meteo['DD']))
df_bagnold = 1.8 * 1.25/9.81 * np.sqrt(d/250e-6) * alpha**3 * (u10 - uth)**3 # kg/m/s

print (df_bagnold*np.maximum(0., df_dir)).ix[t0:t1].sum() * 3600. * 100. / 2650. / .6 / 1e4 # 10^4 m^3, positive alongshore component only
print (df_bagnold*np.minimum(0., df_dir)).ix[t0:t1].sum() * 3600. * 100. / 2650. / .6 / 1e4 # 10^4 m^3, negative alongshore component only
print (df_bagnold*df_dir).ix[t0:t1].sum() * 3600. * 100. / 2650. / .6 / 1e4 # 10^4 m^3, net alongshore component

print 109.815830283 / 4000. * 1e4, 2.66637366864 / 100. * 1e4
```

# Compute transport volumes

```python
def compute_volumes(df):
    V1 = (df*(df_dir>0.)).ix[t0:t1].sum() * 3600. * 4000 / 2650. / .6 / 1e4 # 10e4 m^3, wind energy only
    V2 = (df*np.maximum(0., df_dir)).ix[t0:t1].sum() * 3600. * 4000 / 2650. / .6 / 1e4 # 10e4 m^3, cross-shore component only
    V3 = (df*np.maximum(500.,df_dir*4000.)).ix[t0:t1].sum() * 3600. / 2650. / .6 / 1e4 # 10e4 m^3, minimum orifice of 1 km
    return V1, V3, V2

df_dir = np.maximum(0., np.cos((312.-meteo['DD'])/180.*np.pi))
df_bagnold = 1.8 * 1.25/9.81 * np.sqrt(d/250e-6) * alpha**3 * (u10 - uth)**3 # kg/m/s
df_kawamura = 2.78 * 1.25/9.81 * alpha**3 * (u10 + uth)**2 * (u10 - uth) # kg/m/s
df_horikawa = 1. * 1.25/9.81 * alpha**3 * (u10 + uth)**2 * (u10 - uth) # kg/m/s
df_lettau = 6.7 * 1.25/9.81 * np.sqrt(d/250e-6) * alpha**3 * (u10 + uth) * u10**2 # kg/m/s

volumes = (compute_volumes(df_bagnold),
           compute_volumes(df_horikawa),
           compute_volumes(df_kawamura),
           compute_volumes(df_lettau))

for V in volumes:
    for Vi in V:
        print Vi, 40./Vi
    print ''
```
